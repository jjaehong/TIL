## 1. What is Ethereum?



#### 1. 블록체인과 이더리움의 역사


1. 비트코인 백서(white paper)
    - https://bitcoin.org/bitcoin.pdf 
    - 비트코인 : P2P 거래(송금) 시스템
    - 사람들이 은행 같은 중앙 기관의 필요 없이 서로 직접 돈을 낼 수 있는 시스템을 만들었다고 묘사함
    - 비트코인 블록체인
        - 서로 다른 사람들 사이의 화폐 거래 목록을 저장하는 도구
        - 유일한 목적(결제, 송금, 이체) : 금융 거래를 가능하게 하는 것

2. 이더리움 백서
    - https://web.archive.org/web/20131228111141/http://vbuterin.com/ethereum.html
    - 이더리움 : 스마트 계약과 탈중앙화된 애플리케이션 플랫폼(비탈릭 부텔린)
    - 스마트 계약(smart contract)
        - 이더리움 블록체인에 있는 코드
        - 이 계약서로 어떤 사람이나 다른 계약자가 메시지를 보내도록 할 수 있음
        - 이러한 블록체인에서 작동하는 스마트 컨트랙트로 고급 애플리케이션을 만들 수 있음 


#### 2. 이더리움 네트워크 특징

![이미지](../이미지/이더리움%20네트워크.PNG)

1. 이더리움 네트워크는 노드들 사이에 자금을 이체하거나 데이터를 저장하는 데 사용됨

2. 이더리움 네트워크는 많고 다양하다
    - 테스트 네트워크 : 오직 테스트 코드와 트랜잭션을 테스트하기 위해 사용되는 네트워크

3. 네트워크는 하나 이상의 노드로 형성된다.

4. 각 노드는 이더리움 클라이언트를 실행하는 컴퓨터이다.
    - 노드를 하나 골라서 소프트웨어를 설치하고 실행하면 이더리움 네트워크에 연결되어 전체 네트워크의 일부가 된다. 그리고 이 모든 노드가 연결되고 실제 네트워크를 형성한다.

4. 누구나 노드를 실행할 수 있음
    - 클라이언트 소프트웨어를 다운로드하고 실행하면 노드를 만들고 실행할 수 있음

5. 우리가 만드는 각 노드 또는 네트워크의 각 노드는 블록체인의 완전하고 분리된 복사본을 갖고 있음

6. 블록체인은 데이터베이스이다. 
    - 서로 다른 회사 간의 일어나는 모든 자금 이동의 기록(데이터)을 저장



#### 3. 이더리움 네트워크 연결

1. web3.js 라이브러리
    - for 개발자들
    - 코드를 통해 네트워크와 통신하는 실제 응용 프로그램을 만드는데 사용하는 기술
    - 즉, 웹3를 이더리움 네트워크로 가는 일종의 포털이라고 생각(돈, 데이터 저장, 계약 배포 등)

2. Metamask
    - for 소비자들
    - 메타마스크는 크롬 확장 또는 브라우저 확장으로 일반인들이 이더리움 네트워크와 상호 작용 할 수 있게 해줌 

3. MistBrowser
    - for 소비자들
    - 다양한 이더리움 앱을 찾기 위해 사용되는 완전한 기능을 갖춘 웹 브라우저

#### 블록체인의 활용 사례
    - 블록체인은 중앙화를 대체하는 탈중앙화다.
    - (탈중앙화) 서버를 운영하는 사람들이 독점적으로 업데이트를 막고자함
    - (STO) 자산에 대해 토큰화하여 각 토큰이 소유권을 주장할 수 있게끔 증권사에서 활용 -> 고객들에게 신뢰를 주기 위해 블록체인 활용

    - 블록에 담는 데이터는 수정이 일어나지 말아야 할 민감한 데이터들, 신뢰를 받아야할 데이터들

    - 서버에 담는 데이터는 대용량 데이터나 민감하지 않은 데이터들
