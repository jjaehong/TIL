#### 1. 신입사원 직장생활이란 무엇일끼?

- 직장에서 내 일만 잘하면 되는 것 아닌가?

- 직장이란

    - 성과로 평가 받는 곳(일에 대해 주인의식을 갖고 최선을 다하기)

    - 역할이 주어지는 곳, 책임이 따르는 곳 => 월급을 줌
    
    - 많은 사람들이 함께 일하는 곳(매너 필수)

- 직장생활을 잘 하기 위한 능력

    1. 사람 관계 (상대방에 대한 관심과 애정)

    2. 매너(manner) : 상대방을 존중, 배려하는 마음과 행위

        - 서로의 생각과 입장이 다르기 때문에 이해하고 배려해주는 것이 중요!(이타적인 특성)


- 팔로워십

    - 지위에 의한 갑을 관계가 아니라 역할에 따른 대등한 관계

    - 능동적, 주체적 역할, 정보의 공유

    - 비전을 위해 지식과 아이디어를 쏟아내는 용기 있는 행동(적극적 참여, 질문 및 협업, 높은 성과를 달성하고자 하는 욕구)


- 좋은(GOOD) 직원의 유형

    - 시킨 것 이상으로 더 잘 해내는 직원

    - 시키지 않아도 스스로 찾아서 하는 직원(뭘 해야할지 모르면 물어봄)

    - 판단하기 어려울 때 상의하러 오는 직원

    - 궁금하다고 생각할 때 보고서 들고 오는 직원




#### 2. 상황별 비즈니스 매너 및 커뮤니케이션


1. 직장 내 중요한 매너는 무엇일까?

    - 출퇴근을 비롯 마주쳤을 때 인사하기(52%)

    - 공손한 언어 사용하기(45%)

    - 부르면 대답하고 바라보기(23%)

    - 근무 중 개인적인 일 하지 않기(21%)
    
    - 자기주변 정리정돈(19%)


2. 인사 왜 중요한가?

    - 상대방에 대한 예의와 존중의 표현, 스스로의 이미지를 높이는 기준

    - 기억에 남는 인사? : 밝은 표정, 밝은 음성, 정중한 자세, Small Talk

    - 좋은 인상을 주는 첫걸음은 **인사**입니다. 좋은 인사는 밝은 인사가 아니라 상황에 맞는 인사입니다.


3. 상황별 인사 정리

    - 인사는 누가 먼저 하나요? : 내가 **먼저 !** 인사합니다.

    - 인사 '후'에도 마주친다. 어떻게 해야 할까요? : 가볍게 목례합니다.

    - 화장실에서도 인사 해야 하나요? : 인사 NO, 상대가 인사 시 가볍게 목례합니다.

    - 계단에서 선배나 상사를 마주 쳤을 때 어떻게 해야 하나요? : 기다렸다 인사합니다. < 상대보다 높은 위치에서 인사 NO, 같은 계단에 왔을 때 인사 > 
 

4. 외부 미팅 

    - 서로 알고 있는 가벼운 이야기로 시작해보세요.(스몰 토크 : 아이스 브레이킹)

        - 가벼운 주제로 시작하기 : 식사 / 날씨 / 교통 / 환경 / 일 / 뉴스 / 스타일 / 취미

        - 간단한 질문하기 / 공감하기 / 칭찬하기

    - 두 번째 만남에서는 전에 했던 이야기를 기억해서 시작해보면 좋아요.


5. 명함 매너

    - 인사를 하며 명함을 주고 받는다.

    - 명함을 일어서서 두 손으로 주고 받는다.


6. 외근 시 매너

    1. 외근 시 퇴근 여부 확인

        - 외근 후 퇴근 시간과 맞물렸을 때 퇴근 여부를 직접 판단하지 않습니다.

        - 메시지 보다는 전화로 확인하는 것이 좋습니다. : 저 지금 회사에 들어가려고 하는데, 회사에 들어가면 늦을 것 같은데, 안계시면 어떡하죠?


7. 사내 및 외부 메신저 매너

    - 사내 메신저, 채팅은 업무시간에만 하기!

    - 핵심내용만 정리하여 보내기

8. 이메일 매너

    - 제목을 구체적으로 작성

    - 자신의 소속과 이름을 밝히며 인사한다. 

    - 한눈에 읽기 쉽게 작성한다.

    - 파일첨부가 있는 경우 본문에 표시한다.


9. 재택근무 매너

    - 업무 효율성을 위한 업무 공간 확보

        - 화상 회의 시 복장 갖추어 입기

        - 화상 회의 시 주변 소음, 환경 관리 하기


10. 일 잘하는 질문 매너

    1. 일이 여러가지로 겹치는 상황(일의 우선순위)

        - 우선순위는 스스로 정하지 말고 업무를 지시한 사수에게 문의하도록 합니다.

    2. 자료를 알아서 정리해서 보고했는데, 탐장님께서 본인의 원하는 것이 이것이 아니라며 말을 이해 못했냐고 말함..

        - 평소에 상사가 원하는 방향과 자신이 생각한 방향을 맞추어 보는 것이 중요하다.
        
        - 중간보고 / 이렇게 이렇게 하라는 말씀이 맞으실까요?


11. 업무 핵심 포인트

    1. 지시 받기 전 메모할 준비를 갖춘다.(필수)

    2. 꼭 확인해야 할 지시내용을 **구체적으로 질문**하고 요약하여 확인한다.
        - (왜-무엇-방법-목표를 기준으로 질문을 구성하면 좋음)

    3. 지시사항을 수행하는데 도움이 될만한 참고자료가 있는지 질문함

    4. **마감기한**을 반드시 재 확인하고, 1차 초안검토가 필요할 지 확인함.

    5. 두 건 이상의 지시를 받게 되었을 경우, **우선순위를 스스로 판단하지 않고 상사에게 조언**을 구한다.

    6. 업무 진행 중 조언을 구해야 하는 경우 자신이 생각하고 있는 내용에 2~3개를 뽑아서 이 중 무엇으로 진행해야 하는지 선택형 질문을 함.




#### 3. 탁월한 업무 성과를 위한 비즈니스 커뮤니케이션

1. 비즈니스 커뮤니케이션 왜 중요할까?

    - 조직 : 공동의 목표

    - 공동의 목표 달성을 위한 협업

    - 협업을 이끌어 낼 수 있는 커뮤니케이션

2. 비즈니스 커뮤니케이션

    1. 핵심문장, 단어 반복

        - 상대방에 대한 적극적 관심의 표현임(경청의 증거)

        - 해석상의 오류가 문제를 일으키기 전에 정정 가능함

        - 어떤 정보가 와 닿는지 상대방에게 알려줌

    2. 맞장구

        - 가벼운 맞장구 : 저런, 아 진짜요?, 그렇습니까?, 잘 됐습니다.

        - 동의하는 맞장구 : 과연, 정말 그랬겠군요, 알겠습니다, 레전드레전드

        - 정리하는 맞장구 : 말하자면 이런 것입니까?, 그 말씀은 이것과 저것을 말하고 싶다는 것이지요?

        - 재촉하는 맞장구 : 그래서 어떻게 되었습니까?, A는 그렇게 됐고 B는 어떻습니까?

3. 비즈니스 커뮤니케이션 - 이것만은 방지하자 !

    3. 업무 보고와 업무 협조 요청 시 : 내용을 산발적으로 줄줄 읊지 않도록 주의

## 5. 자기소개서

강연자 : 취업강사(자기소개서/인적성/면접)

#### 1. 채용 트렌드 이해와 채용조건 분석

1. 지원 기회가 많은 SW직군

    1. 플랫폼 기반 SW

        - 네이버 / 카카오 / 라인 / 우아한 형제들

    2. 제조 기반 SW

        - 반도체 / 가전 / 배터리 / 자동차

    3. IT Service 기반 SW

        - 삼성SDS / LG CNS / SK C&C / 현대 오토에버

    4. **금융 기반 SW : 디지털 전환 가속화**

        - 은행 / 증권사 / 보험사 / 카드사


2. 어떻게 해야하는가?

지원회사에 대한 직무 타겟팅 -> 관련 플젝 수행으로 직무역량 채움 / 포트폴리오 정리

직무 타겟팅 및 직접적 직무역량 필요

- 수시채용 전환 및 확대 -> 부서에서 채용하는 데 부서에 fit한 사람 채용
- 실력 중시(코딩 테스트)의 플랫폼 기반 업체의 채용절차

- SW -> 어떤 직무? : 백엔드?, 프론트엔드?, 기획?, 데이터?


- 우대사항


#### 2. 서류전형 및 평가 이해

1. 적정 기준 만족 후 실무진 검토(스펙)

    - 영어성적(토스 IM), 기본 자격증 등

2. 자기소개서 작성 = 면접 준비

    1. **자신감(자존감)**

        - 본인을 설득하지 못하는 지원자는 타인도 설득할 수 없다.

    2. 경험정리

        - 활용할 수 있는 다양한 소재 보유 -> 다각적 면접 질문

    3. 목표분석

        - **회사, 사업부, 직무 이해** -> 지원동기, 입사 후 포부, 직무역량 검증

        - 경험정리와 목표분석의 교집합! => 나의 역량ㆍ경험 == 직무

        - **경험정리 & 목표분석 : 소재 준비 및 선별**

    4. 자기소개서 및 이력서

        1. 이력서 & 자소서 내용 정리 : 면접에서 말할 소재 정리(강점, 보완점)

            - 예시 : 강점(프론트엔드) / 보완점(학점)

        2.  정보 기반의 **가능성 설득**(직무 적합성, 조직 적응성)

        3. 불리한 내용 회피

            - '제가 비록 학점이 낮지만'과 같이 보완의 의도를 보이지 안흥ㅁ
            - 불필요한 내용은 이력서와 자소서에 미기입

        4. 직무 역량의 종류

            - **직접 역량(직무기술서)** : 역량과 직접적으로 연결되는 경험
            - 간접 역량 : 지식 + 유사경험을 통해 연결되는 경험
            - 잠재 역량 : 지원 직무와 연결되는 경험은 아니지만 연결고리를 만들어 낼 수 있는 경험

        5. 직무역량의 질문형태

            1. 경험 시리즈

                - 도전(성공/실패)의 경험에 대해 작성
                - 협력(팀웍,리더십)의 경험에 대해 작성

            2. 대내외 활동

                - Q 학교 생활에 대해 작성
                - Q 대외활동에 대해 작성

            3. 직무 역량

                - Q 직무 전문성을 위해 노력한 경험에 대해 작성 
                - Q 남들과 다른 차별화된 역량에 대해 작성

    5. 태도(예의)

        - 표현의 예의, 자세의 예의

    6. 이미지

        - 시각적, 청각적, 언어적 요소의 작용 -> 호감 포인트

    7. 면접

        - 같이 일할 사람 채용, 종합적 판단

        - 자소서 바탕 정보 기반의 설득과 검증
