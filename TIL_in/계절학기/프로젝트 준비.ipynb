# 프로젝트 준비

2024.01.24

허정욱 프로님

- 프로젝트를 잘 하려면?

    1. **실제 문제 해결** (실용성)
        
        프로젝트가 실제 문제를 해결하거나 유용한 기능을 제공하는 경우 그 가치는 더욱 높아짐
        
    2. **기술적 도전** (학습 능력)
        
        새로운 기술, 프레임워크 또는 기술적으로 도전적인 문제를 해결하는 프로젝트는 기술적 역량을 향상시킴
        
    3. **팀워크 및 협업** (팀 내 작업능력)
        
        타 개발자와 협업 경력은 현업에서 매우 중요
        
        팀워크, 커뮤니케이션, 협업 도구 사용 능력은 취업에도 중요함
        
    4. 완성도
        
        실제로 구현된 아이디어를 보여줌
        
        프로젝트 관리 및 지속능력
        
    5. 문서화 프레젠테이션
        
        아이디어와 기술적 세부사항을 명확하게 전달할 수 있는 능력
        
    6. 사용자 피드백 반영
        
        피드백을 반영하여 제품의 사용성 및 시장적합성을 높이는 데 기여함
        
    7. 지속적인 학습 및 개선
        
        새로운 기술을 익히고 기존 지식을 확장하고, 개선하려는 노력이 필요함
        
    8. 창의성 및 혁신
        
        기존의 방식을 뛰어넘는 새로운 아이디어로 기술적 혁신 가능성을 탐색

        - 기존 대비 차별성 있는 내용과 활용성
