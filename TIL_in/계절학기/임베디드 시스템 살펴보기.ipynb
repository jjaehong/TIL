임베디드 시스템
--------

### 정의

시스템을 동작시키는 소프트웨어를 하드웨어에 내장하여 특수한 기능만을 수행하는 컴퓨터 시스템

### SW의 기본

*   시스템을 구동하기 위한 전용 SW를 포함하고, **부팅 시 자동실행**
*   자동 실행된 SW들을 **모니터링**하고 **이상동작시 문제를 해결**해주는 SW가 탑재

⇒ 요건을 충족하면 임베디드 시스템이라고 할 수 있다(x86 + MS Win 11도)

ex) ATM, KIOSK 등

왜 임베디드 시스템의 UI를 만드는가
--------------------

### UI Application Framework 형태별 특징

*   Native UI Application
    *   C/C++, Java
    *   HW 자원 사용 ***효율성 높음***
        *   높은 속도
        *   적은 메모리 사용량
    *   개발 **난이도 높음**
    *   **개발기간**이 상대적으로 **긺**
    *   개발**비용이 비쌈**
    *   **업데이트 어려움**
*   Web UI Application
    *   HTML5, JavaScript
    *   HW 자원 사용 ***효율성 낮음***
        *   느린 속도
        *   많은 메모리 사용량
    *   개발**난이도** 상대적으로 **낮음**
    *   개발**기간 짧음**
    *   개발비용 상대적으로 **저렴**
    *   **업데이트 쉬움**

### Embedded System용 저전력 Chip 성능 향상

*   기술의 진보로 저가의 저전력 CPU의 대폭적인 성능향상
*   웹 기반의 UI/UX가 시스템에 부하를 주지 않고 충분히 사용 할 만한 시스템이 됨

⇒ Web UI Application의 **HW 자원 효율성이 양호해짐**

→ 웹 기반 상용제품들이 출시됨(스마트 키오스크, 스마트 TV 등)

### Hybrid Application

*   Native UI Application + Web UI Application
*   Native로 만들되, 페이지마다 Native UI와 WebView를 원하는 방식으로 구성 가능하도록 설계
*   Native : 변화가 적거나 빠른 반응속도가 필요한 페이지
*   Webview : 변화가 많거나 빠른 반응속도가 필요 없는 페이지

### 상용 Embedded System에서의 Process Management

*   일반적으로 별도의 프로세스 관리 데몬을 개발하여 관리
*   미들웨어에서 프로세스 모니터링 하는 경우도 있음

→ PM2활용

[PM2 - Home](https://pm2.keymetrics.io/)

정리
--

*   임베디드 시스템은 특수 목적을 위해 제작된 전용 시스템
*   CPU나 OS가 무엇이든 임베디드 시스템의 여부와 관계없음
*   과거에는 HW 성능 문제로 WebApp이 임베디드 시스템에서 작동하기 어려웠음
*   기술의 진보로 저전력 CPU의 성능향상이 있었고, Web기반 임베디드 시스템이 출시되기 시작함
*   WebApp이 활성화된 것은 ‘저렴한 비용’ + ‘상대적으로 쉬운 개발’이 이유
*   Web Front-End, Local Back-End와의 통신은 [Socket.IO](http://Socket.IO) 추천
*   Local Back-End는 node.js나 python, C/C++로 구현 가능
*   프로세스 모니터링 툴은 PM2 추천

웹IoT 기술 기반의 Embedded System 살펴보기
================================

1.  Embedded System이란 무엇인가?
    *   전통적인 임베디드에 대한 인식
        *   물리 장치 만질 것 같음..
        *   기전 시스템..회로.. 칩셋..
    *   임베디드 시스템
        *   시스템을 동작시키는 소프트웨어를 하드웨에 내장하여 `특수한 기능만을 수행`하는 컴퓨터 시스템
        *   `특정 요구 사항`을 명확하게 가지고 있으면 미리 정의된 작업을 수행
        *   예시
            *   라즈베리 파이
                *   라즈베리 파이도 하나의 작은 컴퓨터
                *   라즈베리 파이라고 해서 무조건 임베디드는 아님!
            *   윈도우
                *   위의 요건만 갖추면 윈도우도 임베디드 시스템의 소프트웨어로 작용
                *   ATM은 윈도우를 사용함
    *   임베디드 시스템 소프트웨어의 기본
        *   시스템 구동을 위한 전용 SW
        *   부팅 시 자동실행됨
        *   자동실행되는 소프트웨어를 모니터링하고 이상동작 시 문제 해결하는 소프트웨어 탑재
2.  왜 Web으로 임베디드 시스템의 UI를 만드는가?
    *   UI Application Framework 방식
        *   Native UI App
            *   C or Java
            *   하드웨어 자원 효율 높음
            *   개발 난이도, 비용, 기간 높음
            *   유지보수, 업데이트 어려움
        *   Web UI App
            *   HTML, Javascript
            *   하드웨어 자원 효율 낮음
                *   브라우저와 같은 파서가 필요
            *   개발 난이도, 비용, 기간 낮음
            *   유지보수, 업데이트 쉬움
            *   웹 개발자들도 임베디드할 수 있다~~
    *   기술의 진보
        *   임베용 저전력 칩의 성능 향상
            *   상대적으로 native보다는 밀리지만, 상용화 가능한 수준까지 web ui app의 환경이 갖춰짐
            *   이로부터 기가지니, 스마트 가전 제품이 가능해짐
3.  웹 IoT 프로젝트는 어떻게 만드는 것이 좋은가?
    *   기본 방식
        *   하드웨어 → 디바이스 → OS → 미들웨어 → 어플리케이션
    *   하이브리드 어플리케이션
        *   기반은 native
        *   그러나 페이지마다 native ui와 webview를 원하는대로 구성 가능하도록 설계
        *   변화가 적고 반응 속도가 빨라야 한다면 native
        *   ex) 페이스북, 각종 키오스크
    *   미들웨어 활용
        *   별도의 프로세스 관리 데몬
        *   프로세스를 모니터링하는 미들웨어 활용
        *   PM2
            *   프로세스 모니터링 툴
4.  정리
    *   임베디드는 특수한 목적을 가지고 제작된 전용 시스템
    *   CPS, OS가 무엇이든 임베디드가 될 수 있음
    *   과거에는 상대적으로 개발이 쉬운 WebApp은 작동하기 어려웠으나, 기술의 발전으로 가능
    *   Web Front와 Local Back 간 통신은 [socket.io](http://socket.io) 추천
        *   local back은 node.js나 python, c 사용 가능
    *   프로세스 모니터링 툴은 PM2 추천

*   조언
    *   마크 저커버그
        *   우리에게 가장 큰 위험은 위험을 무릅쓰지 않는 것
        *   급변하는 세상에서 실패할 수밖에 없는 전략은 위험을 감수하지 않는 것
