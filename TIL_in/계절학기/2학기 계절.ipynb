## 1. 웹기술

- WebRTC 화상회의 서비스

- 음성 / 영상 / P2P 서비스

#### 1. 프로젝트 개요

- **WebRTC 화상회의**는 그룹 커뮤니케이션을 위한 그룹 화상과 그룹 채팅 등의 기능이 있는 서비스를 말함.

1. 프로젝트 목표

    - 스켈레톤 코드를 바탕으로 SPA의 기본 틀을 구현

    - Vue3 이해도 향상

    - 로그인, 회원 가입 등 기본 기능 구현

    - 컨퍼런스 생성, 목록, 조회, 수정, 삭제 등의 CRUD 기능 구현

    - WebRTC/WebSocket 기술을 활용한 그룹 화상 회의와 채팅 기능 구현

2. 프로젝트 특징

    - 서비스 기본 구조의 이해
        - 회원가입, 로그인 등 기초 기능을 개발하며 이해도를 향상시킬 수 있습니다.
        - 비디오 컨퍼런스 서비스의 기본 기능인 방 생성, 조회, 상세보기, 회의 이력 조회 기능을 개발해보며 웹 서비스의 기본 구조를 학습할 수 있습니다.

    - WebRTC 기반의 그룹 화상 회의 구현

        - 별도로 제공되는 미디어 서버를 연동하여 WebRTC 기반으로 그룹 화상 회의 기능과 그룹 채팅 기능을 개발해보고 이를 확장하여 화상 기반의 비대면 서비스를 구현할 수 있습니다.

        - WebRTC 및 WebSocket을 통해 영상과 메시지 커뮤니케이션 기능을 구조적으로 설계하여 완성 후 활용도가 높습니다.

        - 기존 화상회의 서비스와 차별점!, 활용도가 높은 서비스, 완성도가 높은 서비스 구현해라 !

#### WebSocket의 이해

1. 개념 : 서버와 클라이언트 간에 **언제든 양방향 통신**이 가능하도록 하는 기술(실시간 통신에 많이 쓰이는 기술)

    ![이미지](../이미지/웹소켓.png)

2. 특징 :

    - 실시간, 양방향 및 지속적인 연결

    - 짧은 지연 시간의 데이터 교환

    - 실시간 채팅, 게임, 주식 차트에 적합

    - websocket emulation(socket.io, sockjs)

3. 자기주도 PJT 목표

    - 웹소켓 개발 환경 구축(Node.js와 Express.js 사용)

        - Node.js + Express.js로 서버 구축

    - 웹소켓을 이용한 메시지 전송 구현(클라이언트에서 서버로)

        - 클라이언트에서 서버로 메시지 전송

        - 버튼 클릭시 서버로 텍스트 메시지 전송(명세서에 나와있음)

    - 양방향 메시지 전송 구현(브로드캐스트)

        - 브로드캐스트(broadcast)사용

        - 매우 간단한 채팅 기능 구현

## 2. 웹디자인

- **웹 큐레이팅 SNS** : "SNS 뉴스피드"는 나의 관심사(키워드)를 바탕으로 웹 페이지를 빠르게 검색하여 유저들과 공유할 수 있는 SNS 기반의 뉴스 큐레이팅 서비스입니다.

    ![이미지](../이미지/6주과정로드맵1.png)

    ![이미지](../이미지/2차%20프로젝트%20진행.png)


#### 1. SUB PJT 1(개인)

- 프로젝트 목표

    - SNS 뉴스 큐레이팅 서비스의 UI/UX를 기획한다.

    - 웹 개발 환경을 구성한다.
    
    - 로그인 단을 개발한다.
    
    - 프론트엔드 / 백엔드 기술의 기본적인 이해도를 함양한다.

#### 2. SUB PJT 2(개발)

- 프로젝트 목표

    - 2차 와이어프레임을 작성한다.
    
    - 프론트엔드, 백엔드 기반의 테스트 코드를 작성한다.
    
    - SNS의 기본 기능들 (뉴스피드, 게시글 작성, 팔로우, 좋아요 등)을 개발한다.


#### 3. SUB PJT 3(open)

- 프로젝트 목표

    - 3차 와이어프레임을 작성한다.
    
    - 웹서비스의 고급 기능들(통계, 트래킹, 필터 및 검색 등)을 개발한다.
    
    - UX 사용성 테스트를 진행하고 개선한다.
    
    - 테스트 코드 작성, 프론트엔드 테스트 자동화, 백엔드 테스트 맛보기, AWS 자동 
    배포
