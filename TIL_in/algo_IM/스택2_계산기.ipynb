 #### 1. 계산기 1

- 문자열로 된 계산식이 주어질 때, 스택을 이용하여 이 계산식의 값을 계산할 수 있다.

- 문자열 **수식 계산**의 일반적 방법

    **step1 : 중위 표기법의 수식을 후위 표기법으로 변경한다.(스택 이용)** 
    
    => 왜? 후위표기법은 괄호를 없애주기에 컴퓨터가 계산하기에 편리해짐
    
    **step2 : 후위 표기법의 수식을 스택을 이용하여 계산한다.**

    - 중위표기법(infix notation)
        - a+b

    - 후위표기법(postfix notation)
        - ab+

#### step1. 중위표기법 -> 후위표기법 **변환**방법 1
- 수식의 각 연산자에 대해서 우선순위에 따라 괄호를 사용하여 다시 표현한다.
- 각 연산자를 그에 대응하는 오른쪽 괄호의 뒤로 이동시킨다.
- 괄호를 제거한다.

#### step1. 중위 표기법 -> 후위 표기법 **변환** 알고리즘(스택이용)2

① 입력 받은 중위 표기식에서 토큰을 읽는다.

② 토큰이 피연산자이면 토큰을 출력한다.

③ 
- 토큰이 연산자(괄호포함)일 때, 이 토큰이 스택의 top에 저장되어 있는 연산자보다 우선순위가 높으면 스택에 push하고,
- 그렇지 않다면 스택의 top의 연산자의 우선순위가 토큰의 우선순위보다 작을 때까지 스택에서 pop한 후 토큰의 연산자를 push한다.
- 만약 top에 연산자가 없으면 push한다.

④ 토큰이 오른쪽 괄호 ')'이면 스택 top에 왼쪽 괄호 '('가 올 때까지 스택에 pop 연산을 수행하고 pop한 연산자를 출력한다. 왼쪽 괄호를 만나면 pop만 하고 출력하지는 않는다.  

⑤ 중위표기식에 더 읽을 것이 없다면 중지하고, 더 읽을 것이 있다면 1부터 다시 반복한다.

⑥ 스택에 남아있는 연산자를 모두 pop하여 출력한다. 

 - 스택 밖의 왼쪽 괄호는 우선 순위가 가장 높으며, 스택 안의 왼쪽 괄호는 우선 순위가 가장 낮다.

```python
# 계산기 1 : 중위표기법 -> 후위표기법(infix => postfix)

# 피연산자 -> 후위표기식에 넣기
# 연산자 -> 스택에 넣은 후, 우선순위대로 빼내고 -> 후위표기식에 넣기

# => 괄호를 모두 없애주기(닫는 괄호면 여는 괄호 나올 때까지 pop)
# => 괄호 안에 연산자들을 먼저 계산할 수 있도록 우선순위를 정해주고 pop함.

# 스택 밖에 있을때 우선 순위
icp = {"+": 1, "-": 1, "/": 2, "*": 2, "(": 3} # in-coming priority # 여는 괄호는 넣을 때 무조건 push라서 우선순위 젤 높음
# 스택 안에 있을때 우선 순위
isp = {"+": 1, "-": 1, "/": 2, "*": 2, "(": 0} # in-stack priority # 여는 괄호는 꺼낼 때마다 버릴것이기에 우선순위 젤 낮음


# infix : 중위표기식
# n : 식의 길이
def get_postfix(infix, n):
    postfix = ""  # 결과로 출력할 후위 표기식
    stack = []

    # 문자열(중위표기식)에서 글자 하나씩 떼오기
    for i in range(n):
        # i번째 글자가 피연산자인지
        if infix[i] not in ["(", "+", "-", "/", "*", ")"]:
            # 결과에 출력
            postfix += infix[i]

        # i번째 글자가 연산자인지
        else:
            # 닫는 괄호가 나오는 경우
            if infix[i] == ")":
                # 여는 괄호가 나올때까지 pop 해서 결과 출력 => pop한 건 후위표기식으로 출력  
                while stack:
                    temp = stack.pop()
                    if temp == "(":
                        break
                    postfix += temp
            # 닫는 괄호가 아닌 다른 연산자가 나오는 경우
            else:
                # 결국 후위표기식의 연산자들은 stack에 넣어진 후 pop하고 postfix에 넣어짐
                # 후위표기식을 계산하는데 우선순위 높은 연산자들을 먼저 계산해야하므로, 스택에서 더 빨리 빼내야함.(괄호 고려)
                # 토큰 연산자와 스택 안의 연산자를 비교하여
                # 스택 안의 연산자가 우선순위가 높거나 같으면 꺼내서 후위표기식에 추가함
                # infix[i]가 우선순위가 낮으면 스택에 넣음
                # 따라서 스택에는 뒤에서부터 비교적 우선순위가 높은 연산자로 채워지게 됨. 왜? 우선순위 높은 것을 먼저 출력해야하니까!
                # 따라서, stack[top]에 있는 연산자의 우선순위가 현재 토큰(연산자)의 우선순위보다 같거나 높으면 (계속 비교)
                while stack and isp[stack[-1]] >= icp[infix[i]]:
                    postfix += stack.pop() # 계속 pop 해서 출력
                # 그게 아니면 스택에 push()
                stack.append(infix[i])

    # 닫는 괄호까지 출력하고 남아있는 연산자들 출력
    while stack:
        postfix += stack.pop()

    return postfix
```

#### 2. 계산기 2
- step2. 후위 표기법의 수식을 스택을 이용하여 **계산**

① 피연산자를 만나면 스택에 push한다. 

② 연산자를 만나면 필요한 만큼의 피연산자를 스택에서 pop하여 연산하고, 연산결과를 다시 스택에 push한다.

③ 수식이 끝나면, 마지막으로 스택을 pop하여 출력한다.

```python
# forth 
# 계산기 2 : 후위표기법 -> 계산

# 피연산자 -> 스택
# 연산자 -> 스택에서 피연산자 꺼내서 계산 실행

# 후위표기법의 결과를 계산할 함수
def get_result(postfix):
    stack = []

    # 후위표기식에서 글자 하나씩 떼오기
    for c in postfix:

        # '.'을 만나면 숫자 출력
        if c == '.':
            if len(stack) == 1:
                return stack.pop()
            else:
                return 'error'
        
        # 피연산자를 만나면 스택에 넣기
        elif c not in '+-/*':
            stack.append(int(c)) # 타입 조심
        
        # 연산자를 만나면 스택에서 연산에 필요한 만큼 꺼내서 계산
        # 계산한 결과를 다음 연산에 쓰기 위해서 push
        else:
            if len(stack) >= 2:

                # 오른쪽이 먼저
                right = stack.pop()
                # 왼쪽이 나중에
                left = stack.pop()

                # 연산자의 종류에 따라서 계산
                if c == '+':
                    result = left + right
                elif c == '-':
                    result = left - right
                elif c == '/':
                    result = left // right
                elif c == '*':
                    result = left * right
            else:
                return 'error'
            
            # 계산 결과를 다음 연산자에서 쓰기 위해
            # 여기의 계산 결과가 다음 연산자의 피연산자가 됨
            stack.append(result)

     # 마지막에 남은 결과 하나 꺼내서 확인
    return stack.pop()


T = int(input())
for tc in range(1,T+1):
    line = input().split()

    print(f'#{tc} {get_result(line)}')

```

```python
# 재귀함수란
# 재귀함수는 자신의 로직을 내부적으로 반복하다가, 일정한 조건이 만족되면 함수를 이탈하여 결과를 도출함.

def f(i,n):
    
    # 1. 종료조건
    if i == n:
        return
    
    # 2. 재귀호출
    else:
        B[i] = A[i]
        f(i+1, n)
        return

N = 3
A = [1,2,3]
B = [0] * N
f(0,N)
print(B)
```

```python
# 부분집합 구하기(재귀함수 사용)

# i번째 원소를 부분집합에 포함할지 안할지를 결정해준다.
# n-1 번째 원소까지 하면 된다.
# n-1 번째 원소까지 완료한 경우 뒤로 돌아와서 내가 이전에 선택했다면
# 선택하지 않고 진행 ==> 반복


numbers = [1,2,3,4]
selected = [0] * 4
# selected[i] == 1 : 내가 i번째 원소를 부분집합에 포함 o
# selected[i] == 0 : 내가 i번째 원소를 부분집합에 포함 x
n = len(numbers)

def subset(i,n):
    # 1. 종료 조건
    if i == n: 
        # n개의 원소에 대한 선택을 끝냈다.(부분집합에 넣던지 안넣던지, n-1번째 원소까지 선택했다.)
        for j in range(n):
            if selected[j]:
                print(numbers[j], end = ' ')
        print()
        return
    
    # 2. 재귀호출
    else:
        # i번째 원소를 선택하고 다음 i+1 번째 원소를 선택하러 가거나
        selected[i] = 1
        subset(i+1, n)

        # i번째 원소를 선택하지 않고 다음 i+1 번째 원소를 선택하러 가거나
        selected[i] = 0
        subset(i+1, n)
        return

subset(0,n)
# 결과 리뷰
# 1~4를 모두 선택한 것을 먼저 출력
# 뒤로 돌아와서 1~3

```

```python
# 부분집합의 합(재귀함수 사용)

numbers = [1,2,3,4,5]
selected = [0] * 5
n = len(numbers)

# 만약 합이 x보다 작은 부분집합만 구해야 한다면?
x = 6

def subset(i,n,total):

    # 1 종료 조건
    if i == n:
    
        # n개의 원소에 대한 선택을 끝냈다.
        temp = 0 
        for j in range(n): 
            if selected[j]:
                temp += numbers[j]
        
        # 합이 x이하인 부분집합만 출력
        if temp <= x:
            for j in range(n):
                if selected[j]:
                    print(numbers[j], end = ' ')
            print(f' : {temp}') # 부분집합 합 한 개씩 표현
        return
    
    
    # 2. 재귀호출
    else:
         # i번째 원소를 선택하고 다음 i+1 번째 원소를 선택하러 가거나
        selected[i] = 1
        subset(i+1, n, total + numbers[i]) # i번째 원소를 선택했기에 subsum에 더해줘야한다.
        # i번째 원소를 선택하지 않고 다음 i+1 번째 원소를 선택하러 가거나
        selected[i] = 0
        subset(i+1, n, total)

subset(0,n,0)
```

```python
# 부분집합의 합(재귀함수 사용)
# + 최적화

numbers = [1, 2, 3, 4, 5]
selected = [0] * 5
n = len(numbers)

# 만약 합이 x보다 작은 부분집합만 구해야 한다면??
x = 6

# 내가 i번째 원소를 선택 or 선택 x 했던 상황에서 합을 기억
def subset(i, n, subsum):

    # 0 : 다른 조건(최적화)
    if subsum > x:
        return

    # 1. 종료 조건
    if i == n:
        
        # n개의 원소에 대한 선택을 끝냈다.(부분집합에 넣던지 안넣던지)
        temp = 0
        for j in range(n):
            if selected[j]:
                temp += numbers[j]

        # 합이 x 이하인 부분집합만 출력
        if temp <= x:
            for j in range(n):
                if selected[j]:
                    print(numbers[j], end=" ")
            print()

        return

    # 2. 재귀호출
    # i번째 원소를 선택하고 다음 i + 1 번째 원소를 선택하러 가거나
    selected[i] = 1
    subset(i + 1, n, subsum + numbers[i])
    # i번째 원소를 선택하지 않고 다음 i+1 번째 원소를 선택하러 가거나
    selected[i] = 0
    subset(i + 1, n, subsum)


subset(0, n, 0)
```

```python
# [참고] 순열

# P[0]결정 -> P[1]결정 -> P[2]결정 -> 완성
# 자리를 바꾸면서 경우의 수 

def f(i, N):
    if i == N:
        print(A)
    
    else:
        for j in range(i,N): # 자신부터 오른쪽 끝까지, 왼쪽으로도 가면 중복되기에 
            A[i], A[j] = A[j], A[i]
            f(i+1, N)
            A[i], A[j] = A[j], A[i]


A = [1,2,3]
f(0, 3)
```

```python
# [참고] 순열(재귀함수)

numbers = [1, 2, 3]
n = len(numbers)

# i번째 원소의 자리를 바꿔가면서 순열 생성
# 자리를 바꿀수 있는 경우의 수
def perm1(i):

    # 1. 종료 조건
    if i == n:
        print(numbers)
        return

    # 2. 재귀 호출
    # 현재위치 i에서 다른위치 j에있는 숫자와 자리를 바꾼다.
    # j를 선택할 때는 중복을 방지하기 위해 i보다 뒤에있는 원소만 선택
    # i,j 가 같을수도 있다. (자리를 안바꾸는 경우)
    # i번째 원소를 누구랑 바꿀지 정했다면 i+1번째 원소를 바꾸러 진행

    for j in range(i, n): # 자신부터 오른쪽 끝까지, 왼쪽으로도 가면 중복되기에 
        # i번째와 j번째 위치를 바꾸고 진행
        numbers[i], numbers[j] = numbers[j], numbers[i]

        # 다음 i+1 번째 원소의 자리를 바꾸러 간다
        perm1(i + 1)

        # i번째와 j번째 원소를 되돌려놓고 다음 진행
        numbers[i], numbers[j] = numbers[j], numbers[i]

perm1(0)
```

```python

numbers = [1, 2, 3, 4, 5]
n = len(numbers)


# 순열의 i번째 자리를 누구로 할것인가 직접 선택하는 방법
# i번째 자리를 누구로 선택했는지 기억해야 중복선택 xx

# selected[i] : i번째 위치에 사용한 원소 (selected 자체가 순열)
# selected = [0] * n

def perm2(i, selected):
    # 종료 조건
    if i == n:
        print(selected)
        return

    # 재귀 호출
    # 현재 위치 i에 누구를 놓을것인가 선택
    # 이전에 선택했던 원소는 선택 xx
    for j in range(n):
        # j번째 원소를 선택한적이 없다면 j번째 원소를 순열의 i 위치에 놓기
        if numbers[j] not in selected:
            # i번째 위치에 j를 놓기
            selected[i] = numbers[j] # 숫자에 중복이 없다는 가정하에 진행
            # i번째 위치에 누굴 놓을지 정했으니 i+1번째에 누굴 놓을지 정하러 가기
            perm2(i + 1, selected)
            # i번째 위치에 j를 놓았던 거를 없던일로 하고 다른 j를 선택하러 간다
            selected[i] = 0


perm2(0, [0] * 5)
```
