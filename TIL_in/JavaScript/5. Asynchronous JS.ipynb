## 5. Asynchronous(비동기) JavaScript

#### 1. 비동기


1. 개요

    - 동기(Synchronous)

        - 프로그램의 실행 흐름이 **위에서부터 아래로 순차적**으로 진행

        - 하나의 작업이 완료된 후에 다음 작업이 실행되는 방식

        ![이미지](../이미지/동기.PNG)

    - 비동기(Asynchronous)

        - 프로그램의 실행 흐름이 순차적이지 않으며, 작업이 완료되기를 **기다리지 않고 다음 작업이 실행**되는 방식

        - 작업의 완료 여부를 신경 쓰지 않고 **동시에 다른 작업들을 수행할 수 있음**

        - **병렬적 수행(당장 처리를 완료할 수 없고 시간이 필요한 작업들은 별도로 요청을 보낸 뒤 응답이 빨리오는 작업부터 처리**)

        - 예시 2 : 네이버 웹 페이지 들어갈 때

            - 화면이 위에서부터 아래로 출력되어 보여주는 것이 아닌

            - 다른 것들보다 이미지들은 늦게 나타남

        - 예시 4 :

            ![이미지](../이미지/비동기.PNG)

#### 2. JS와 비동기

1. 개념

- JavaScript : single Thread 언어

    - Thread란? : 작업을 처리할 때 실제로 작업을 수행하는 주체, multi-thread라면 업무를 수행할 수 있는 주체가 여러 개라는 의미

    - 따라서 JS는 single Thread 언어이기에 동시에 여러 작업을 처리할 수 없음(**하나의 작업을 요청 순서대로만 처리 O**)

    - 그러면 JS는 어떻게 비동기 처리를 할 수 있을까? => "환경(브라우저, node)안에서 가능" 


- JS runtime(런타임) : JS가 실행되는 환경(런타임)


2. 브라우저 환경에서의 JS 비동기 처리 관련 요소

    - JS 엔진의 **Call Stack**

        - 요청이 들어올 때마다 **순차적으로 처리하는 Stack(후입선출)**
        - **기본**적인 Js의 single Thread 작업 처리

    - **Web API**

        - JS 엔진이 아닌 **브라우저에서 제공하는 runtime 환경**
        - **시간이 소요되는 작업**을 처리(setTimeout, DOM Event, AJAX 요청 등)

    - **Task Queue**

        - 비동기 처리된 Callback 함수가 **대기하는 Queue(선입선출)**

    - **Event Loop**

        - 작업이 들어오길 기다렸다가 작업이 들어오면 이를 처리하고, 처리할 작업이 없는 경우에는 잠드는, 끊임없이 돌아가는 JS 내 루프
        - Call 스택과 Task 큐를 지속적으로 모니터링
        - **Call 스택이 비어 있는지 확인 후 비어있다면, Task 큐에서 대기 중인 오래된 작업을 Call 스택으로 Push**

5. JS와 비동기 정리

    - **JS**는 한 번에 하나의 작업을 수행하는 Single Thread 언어로 **동기적 처리를 진행**

    - 하지만 **브라우저 환경에서는** Web API에서 처리된 작업이 지속적으로 Task 큐를 거쳐 Event Loop에 의해 Call 스택에 들어와 순차적으로 실행됨으로써 **비동기 작업이 가능한 환경이 됨**



#### 3. AJAX (Asynchronous JavaScript + XML)

1. AJAX

    - JS의 비동기 구조와 XML 객체를 활용해 비동기적으로 서버와 통신하여 **웹 페이지의 일부분만을 업데이트**하는 웹 개발 기술(**새로고침(전체 페이지 재업로드) 없이도 메서드 요청이 반영됨**)

    - 'X'가 XML을 의미하기는 하지만, 요즘은 더 가벼운 용량과 JavaScript의 일부라는 장점 때문에 **JSON**을 더 많이 사용


2. XMLHttpRequest 객체(XHR 객체) 

    - 서버와 상호작용할 때 사용하며 **페이지의 새로고침 없이도** URL에서 데이터를 가져올 수 있음

#### 4. Callback 과 Promise

0. 들어가기 전

- 비동기 처리의 핵심은 Web API로 들어오는 순서가 아니라 **작업이 완료되는 순서에 따라 처리**한다는 것

- 그런데 이는 코드의 실행 순서가 불명확하다는 단점이 존재함(어떤 api를 가져오느냐에 따라 처리순서가 달라짐 -> 원하는 순서나 로직이 있을텐데, 순서가 꼬일 수 있음)

- 이러한 문제를 콜백 함수로 해결하자!


1. 비동기 콜백(callback)

    1. 개념

    - 콜백함수는 비동기적으로 처리되는 작업이 완료되었을 때 실행되는 함수

    - **연쇄적으로 발생하는 비동기 작업을 순차적으로 동작할 수 있게 하는 로직**(작업의 순서와 동작을 제어하거나 결과를 처리하는 데 사용)


    2. 한계 : 콜백 지옥

    ![이미지](../이미지/콜백지옥.PNG)

    - 비동기 콜백 함수는 어떤 기능의 실행결과를 받아서 다른 기능을 수행하기 위해 많이 사용됨, 이 과정을 작성한 것이 위의 결과

    - 코드의 가독성을 해치고 유지 보수가 어려워짐


2. 프로미스(Promise)

    ![이미지](../이미지/비동기콜백,promise.PNG)
    
    1. 개념

    - promise 또한 연쇄 비동기 작업을 **순차적으로 처리 동작 !**

    - **콜백 지옥 문제를 해결**하기 위해 등장한 비동기 처리를 위한 객체

    - "작업이 끝나면 실행시켜 줄게" 라는 약속(promise)

    - Promise 기반의 클라이언트가 바로 이전에 사용한 Axios 라이브러리

        - 성공에 대한 약속 : **then()**
        - 실패에 대한 약속 : **catch()**
