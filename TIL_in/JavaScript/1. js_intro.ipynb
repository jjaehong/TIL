## 1. Intro_javascript

#### 1. History of JavaScript

1. 웹의 탄생

    - 팀 버너스리 경이 WWW.하이퍼 텍스트 시스템 고안

    - URL, HTTP 최초 설계, 초기의 웹은 **정적인 텍스트 페이지**만을 지원


2. JavaScript 탄생

    - 웹의 동적 기능 개발을 넘어 스크립트 언어 Mocha 개발 -> JavaScript로 이름을 변경

    - JS는 **웹 페이지에 동적 기능을 추가**하는 데 사용됨


3. JavaScript 파편화

    - 많은 기업들이 자체적으로 JavaScript를 독자적으로 변경하고 이를 자체 브라우저에 탑재


4. 1차 브라우저 전쟁(Netscape vs Microsoft)

    - Internet Explorer(IE) 승리

    - 웹 표준의 부재로 인해 각 기업에서 자체 표준을 확립하려는 상황이 벌어짐
    - 이는 웹 개발자들에게 큰 혼란을 주었으며, 결국 웹 표준의 중요성 인식하는 계기가 됨

    - **ECMAScript** 출시 : JavaScript의 파편화를 막기위해 ECMAScript 표준 언어 정의


5. 2차 브라우저 전쟁(IE, Firefox, Chrome)

    - Chrome 승리

    - 웹 표준을 준수하는 Chrome의 등장으로 웹 표준의 중요성이 대두

    - 웹 기능이 크게 확장되며 웹 애플리케이션의 비약적인 발전을 이끌어감

    - **웹의 기술적 발전과 웹 표준의 중요성**

6. **ECMAScript와 JavaScript**

    - ECMAScript
        
        - 표준화된 스크립트 프로그래밍 언어 명세

        - 스크립트 언어가 준수해야하는 규칙, 세부사항 등을 제공

        - ECMAScript는 JavaScript의 표준이며, JavaScript는 ECMAScript 표준을 따르는 구체적인 프로그래밍 언어

        - ECMAScript는 언어의 핵심을 정의하고, JavaScript는 ECMAScript 표준을 따라 구현된 언어로 사용됨

        - ES6

    - JavaScript

        - 현재는 Chrome, Safari, Microsoft Edge 등 다양한 웹 브라우저가 출시되어 있으며, 웹 브라우저 시장이 다양화 되어있음

        - 기존의 javascript는 브라우저에서만 웹 페이지의 동적인 기능을 구현하는 데에만 사용되었음
            
            - 예를 들어, 사용자의 입력에 따라 웹 페이지의 내용이 동적으로 변경되거나, 애니메이션 효과가 적용되는 등의 기능

        - 그러나 이후 브라우저에서 벗어나 Node.js와 같은 서버 사이드 분야 뿐만 아니라, 다양한 프레임 워크와 라이브러리가 개발되면서 **웹 개발 분야애서는 필수적인 언어로 자리 잡게 됨**

    - Script Language
        
        - 자바스크립트도 스크립트 언어의 한 종류

        - 특징 : 프로그램이 실행되는 런타임에 코드가 해석된다는 점(컴파일이 아닌)
