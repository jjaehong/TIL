## Controlling EVENT

#### 1. 이벤트


1. event

    - 일상에서의 이벤트처럼 웹에서도 **이벤트를 통해 특정 동작을 수행**한다.

    - event : 무언가 일어났다는 신호, 사건

        - event 종류 : mouse, input, keyboard, touch ...

    - event object : DOM에서 이벤트가 발생했을 때 생성되는 객체

    - DOM 요소는 event를 받고, 받은 event를 '처리'(**event handler(함수)**)할 수 있음


2. event handler

    - event handler : 이벤트가 발생했을 때 실행되는 함수(**사용자의 행동에 어떻게 반응할지**를 JS코드로 표현한 것)

    - .addEventListener 개념

        ![이미지](../이미지/addeventlistener.PNG)

        - **DOM 요소에서 수신할 이벤트가 실행됐을 때, 이 콜백함수(적용될 대상을 넣어놔)를 부착해서 호출해라!**

        - type : 수신할 이벤트 이름(문자열로 작성, (ex) 'click')

        - handler : 콜백함수는 발생한 event object를 유일한 매개변수로 받음

    - .addEventListener 활용

        ![이미지](../이미지/addeventlistener활용.PNG)

        - console.log(event) : 개발자 도구에서 event 객체의 모든 속성을 확인할 수 있음!

        - 요소에 addEventListener를 부착하게 되면 **내부의 this 값**은 **대상 요소**를 가리키게 됨(event 객체의 currentTarget 속성 값과 동일)

    - addEventListener의 콜백 함수 특징

        ![이미지](../이미지/콜백함수특징.PNG)

3. 버블링(Bubbling)

    ![이미지](../이미지/버블링.png)
    
    - 한 요소에 이벤트가 발생하면, 이 요소에 할당된 핸들러가 동작하고, 이어서 부모 요소의 핸들러가 동작하는 현상

    - 버블링 방향 : 최하위 선택된 요소부터 시작해 부모(최상위요소)로 전파된다.(== 안에서부터 밖으로)

    - 버블링 예시 : 가장 안쪽의 <p>요소를 클릭하면 p -> div -> form 순서로 3개의 이벤트 핸들러가 동작

    ![이미지](../이미지/버블링%20예시.PNG)

    - **이벤트가 정확히 어디서 발생했는지 접근할 수 있는 방법**

        1. event.target

            - 이벤트가 발생한 가장 안쪽의 요소(target)을 참조하는 속성

            - **실제 이벤트가 시작된 target 요소**

            - 버블링이 진행되어도 변하지 않음

        2. event.currentTarget

            - '현재' 요소(보통 최상위 요소까지 버블링되어 핸들러를 실행시키기에..)
            
            - 항상 이벤트 핸들러가 연결된 요소만을 참조하는 속성

            - 'this'와 같음

    - 버블링 막기 : stopPropagation
