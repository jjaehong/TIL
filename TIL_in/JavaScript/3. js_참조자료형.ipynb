3. **this** keyword

    - this : 함수나 메서드를 **호출한 객체**를 가리키는 키워드

    - **this** 키워드를 사용해 **상위 함수 내에서** 객체의 속성 및 메서드에 접근하기 위해 사용

    - **this를 안쓴다는 것은 외부에서 파라미터를 받아와야한다. this를 쓴다는 것은 상위 함수(or 객체)에서 속성을 받아오겠다!!**

    - this 호출 방법

        1. 단순 호출 : **전역 객체(window)** / document ⊂ window

            ![이미지](../이미지/this단순호출.PNG)


        2. 메서드 호출 : 메서드를 호출한 **객체** 

            ![이미지](../이미지/this메서드호출.PNG)


        3. 중첩된 함수에서의 this 문제점과 해결책

            ![이미지](../이미지/this.PNG)

            - #### **화살표 함수의 this는 나를 호출한 함수를 따라가!**

            - 화살표 함수가 가리키는 this는 나를 호출한 함수의 상위 객체

            - 객체의 메서드에서는 화살표 함수를 사용하지않는다.(그 this는 window를 가리키기에)

            - JS 'this' 정리

                - JS에서 this는 함수가 "호출되는 방식"에 따라 결정되는 현재 객체를 나타냄

                - JS의 함수는 호출될 때 **this를 암묵적으로 전달받음**

                - python의 self와 java의 this가 선언 시 값이 이미 정해지는 것에 비해 **JS의 this는 함수가 호출되기 전까지 값이 할당되지 않고 호출 시에 결정**됨(동적 할당, window 인지 객체 인지)

            - JS 'this' 장단점

#### 4. 참고

1. 콜백함수 구조를 사용하는 이유

    1. 함수의 재사용성 측면

        - 따로 관리하다보니 콜백 함수의 동작을 자유롭게 변경 가능

        - 콜백함수를 호출하는 forEach 함수를 호출하는 코드는 간결하고 가독성이 높아짐

    2. 비동기적 처리 측면

        - setTimeout : 콜백함수를 인자로 받아 일정 시간이 지난 후에 실행됨

        - 비동기적으로 콜백함수를 실행하므로 다른 코드의 실행을 방해하지 않음


2. 배열은 객체다

    - 배열은 키(index)와 속성을 담고 있는 참조 타입의 객체

    - 순서가 있는 컬렉션을 제어하게 해주는 특별한 메서드 제공

    - 배열의 키 : index('숫자형 키') 

    - length 속성
