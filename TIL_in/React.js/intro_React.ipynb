## 1. React intro

#### 1. React 개념 및 특징

1. React.js

- 사용자 인터페이스를 만들기 위한 자바스크립트 **라이브러리**


  - 사용자 인터페이스 : 사용자와 컴퓨터 프로그램이 서로 상호작용을 하기 위해 중간에서 서로 간에 입력과 출력을 제어
    
    - 예시 : 버튼을 누름으로써 웹사이트에서 반응을 하고 웹사이트에서 특정 페이지를 보여줌으로써 사용자들이 반응하는 것

  - 라이브러리 : 자주 사용되는 기능들을 모아놓은 모음집(객체에 대한 제어권이 개발자에게 있음)

- 리액트는 SPA를 쉽고 빠르게 만들 수 있도록 해주는 도구임

2. React 장단점

  1. 장점

      1. 빠른 업데이트 & 렌더링 속도

        - virtual DOM :  업데이트해야 할 최소한의 부분만을 찾아서 업데이트 하기에     **업데이트를 빠르게 해줌**. -> 렌더링 속도가 빠르니 유저에게 좋음


      2. 컴포넌트 기반의 구조(구성요소)

        - 레고 블록 조립하듯 컴포넌트들을 모아서 개발

        - 재사용성(Reusability) : 다시 사용이 가능한 성질

            - 개념 : 소프트웨어를 개발할 때, 재사용성을 높게 개발해야한다는 말의 의미   (의존성 낮게, 호환성 높게) : 해당 소프트웨어의 또는 모듈이 다른   프로그램에서도 쉽게 곧바로 쓸 수 있도록 개발하는 것을 의미함

            - 재사용점 장점 : 개발기간이 단축됨 / 유지 보수가 용이함 / 버그 찾기 쉬움

      3. 활발한 지식공유 & 커뮤니티

        - 깃허브 리액트 프로젝트 페이지 / 스택 오버 플로우

      4. React Native

        - 리액트를 배운 이후 react native 프레임워크를 사용해서 모바일 앱(안드로이드 앱, IOS앱)도 개발 가능

  2. 단점

      1. 방대한 학습량

      2. 계속 뭔가 바뀜.. 또 새로 공부해야함

      3. 높은 상태(state)관리 복잡도

#### 2. TypeScript

1. 개념

  - JS를 기반으로 **정적 타입 문법**을 추가한 프로그래밍 언어이다.

  - JS는 동적 타입의 인터프리터 언어로 런타임에서야 오류를 발견할 수 있었다. 하지만 타입스크립트는..

  - TypeScript는 코드 작성 단계에서 타입을 체크해서 오류를 확인할 수 있기 때문에 실행하기 전에 많은 오류들을 잡아낼 수 있는 장점이 있다.

  - 프론트엔드 개발자는 반드시 익혀야 하는 기술 스택!
  

2. 특징
  - **높은 생산성!!**

  - "아니, ctrl + space를 하면 어시스트가 된다고?"

  - "브라우저에서 실행을 안했는데 오류를 찾아낸다고?"

  - "이 족보없는 언어(js)에 족보가 생겼다고?"

  - Type을 개발 단계에서 알 수 있다.

  - low level의 엄격함을 JS에도 도입

  - 100%의 호환성, 빠른 학습

#### 3. Next.js

1. 개념

  - React에 Server Side Rendering(SSR)을 가능하게 해주는 라이브러리

  - CSR(SPA)은 대부분의 검색 엔진들이 CSR로 개발된 페이지를 제대로 인식하지 못하는 단점이 있다. 이는 인터넷 이용하는 사람들에게 검색 상단에 노출되지 않게 되므로 마케팅 비용 증가로 이어짐. 따라서 Next.js로 대체방법 활용


- Next.js 설치

    - npx create-next-app@latest --ts

    - npm run dev
