#### CSS box model

- 모든 html 요소를 **사각형 박스**로 표현하는 개념

    -> **내용(content), 안쪽 여백(padding), 테두리(border), 외부 간격(margin)으로 구성되는 개념**

![이미지](../이미지/box구성.PNG)

```python
<!-- 원 만들기 -->
<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<meta http-equiv="X-UA-Compatible" content="IE=edge" />
		<meta name="viewport" content="width=device-width, initial-scale=1.0" />
		<title>Document</title>
		<style>
			.circle {
				width: 100px;
				height: 100px;
				border: 2px solid black;
				background-color: green;

				/* 테두리 동그랗게 깎기 */
				border-radius: 100px;
			}
		</style>
	</head>
	<body>
		<div class="circle"></div>
	</body>
</html>
```

```python
<!-- 박스 구성요소 예시 -->
<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<meta http-equiv="X-UA-Compatible" content="IE=edge" />
		<meta name="viewport" content="width=device-width, initial-scale=1.0" />
		<title>Document</title>
		<style>
			.box1 {
				width: 200px;
				padding-left: 25px;
				padding-bottom: 25px;
				margin-left: 25px;
				margin-top: 50px;
				/* border-width: 3px;
				border-style: solid;
				border-color: black; */
				border: 3px solid black;
			}

			.box2 {
				width: 200px;
						/*상하 좌우*/
				padding: 25px 50px;
				margin: 25px auto; /*(상하)앞에 div로 부터 25px, (좌우)자동 가운데 정렬*/
				border: 1px dashed black;
			}
			
			body {
				border: 1px solid black;
			}

		</style>
	</head>
	<body>
		<div class="box1">box1</div>
		<div class="box2">box2</div>
	</body>
</html>
```

#### box-sizing

- 박스의 너비나 높이를 정확한 수치로 맞춰주기 위한 속성

![이미지](../이미지/boxsizing.PNG)

```python
<!-- box-sizing -->
<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<meta http-equiv="X-UA-Compatible" content="IE=edge" />
		<meta name="viewport" content="width=device-width, initial-scale=1.0" />
		<title>Document</title>
		<style>
			.box {
				width: 100px;
				border: 2px solid black;
				padding: 10px;
				margin: 20px;
				background-color: yellow;
			}

			.content-box {
				box-sizing: content-box;
			}

			.border-box {
				box-sizing: border-box;
			}
		</style>
	</head>
	<body>
		<div class="box">box</div>
		<div class="box content-box">content-box</div>
		<div class="box border-box">border-box</div>
	</body>
</html>
```

#### 기타 display 속성

1. inline-block
/* display: inline-block; */

![이미지](../이미지/인라인블록.PNG)

- span이 다른 요소를 밀어낼 수 있게 됨
- 리스트 요소가 가로로 정렬이 됨
- div 요소가 가로로 정렬이 됨

2. none

/* display: none; */
- 요소를 화면에서 숨기는데 공간도 부여하지 않아

3. hidden

/* visibility: hidden; */
- 숨기는데 차지하고 있는 공간은 그대로 있어 */

```python
<!-- block-inline -->
<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<meta http-equiv="X-UA-Compatible" content="IE=edge" />
		<meta name="viewport" content="width=device-width, initial-scale=1.0" />
		<title>Document</title>
		<style>

			/* 인라인요소 */
			a, span, img {
				border: 3px solid red;
				/* width: 50px; */
			}


			/* 블록요소 */
			h1, p, div {
				/* width: 300px; */
				border: 1px solid blue;
			}
		</style>
	</head>
	<body>
		<h1>Normal flow</h1>
		<p>Lorem ipsum dolor sit amet consectetur adipisicing elit.</p>
		<div>
			<p>
				block 요소는 기본적으로 부모 요소의 너비 100%를 차지하며, 자식 콘텐츠의
				최대 높이를 취한다.
			</p>
			<p>
				block 요소의 총 너비와 총 높이는 content + padding + border
				width/height다.
			</p>
			<p>block 요소는 서로 margin 으로 구분된다.</p>
			<p>
				inline요소는 <span>이 것처럼</span> 자체 콘텐츠의 너비와 높이만
				차지한다. 그리고 inline 요소는
				<a href="#">width나 height 속성을 지정 할 수 없다.</a>
			</p>
			<p>
				물론 이미지도 <img src="#" /> 인라인 요소다. 단, 이미지는 다른
				inline요소와 달리 width나 height로 크기를 조정할 수 있다.
			</p>
			<p>
				만약 inline 요소의 크기를 제어하려면 block 요소로 변경하거나
				inline-block 요소로 설정해줘야 한다.
			</p>
		</div>
	</body>
</html>
```

## CSS layout position

#### 1. CSS layout 

- 각 요소의 **위치**와 **크기를 조정**하여 웹페이지의 디자인을 결정하는 것
    
    -> display, position, float, flexbox 등

#### 2. CSS position

- 요스를 normal flow에서 제거하여 다른 위치로 배치하는 것
    
    -> 다른 요소 위에 올리기, 화면 특정 위치에 고정시키기 등

- position 유형별 특징
    
    1. static
        - 기본 값
        - 요소를 normal flow에 따라 배치
    
    2. relative
        - 요소를 normal flow에 따라 배치
        - 자기 자신을 기준으로 이동(왼쪽 위의 점이 기준)
        - 요소가 차지하는 공간은 static일 때와 같음

    3. absolute
        => 앱솔루트는 집나간 애다
        - 요소를 normal flow에서 *제거*
        - static이 아닌 가장 가까운 relative 부모 요소를 기준으로 이동
        - 문서에서 요소가 차지하는 공간이 없어짐

    4. fixed
        - 요소를 normal flow에서 제거
        - 현재 화면영역(viewport)을 기준으로 이동
        - 문서에서 요소가 차지하는 공간이 없어짐
    
    5. sticky
        - 요소를 normal flow에 따라 배치
        - 요소가 일반적인 문서 흐름에 따라 배치되다가 스크롤이 특정 임계점에 도달하면 그 위치에서 고정됨(fixed)
        - 만약 다음 sticy 요소가 나오면 다음 sticky 요소가 이전 sticky 요소의 자리를 대체
            - 이전 sticky 요소가 고정되어 있던 위치와 다음 sticky 요소가 고정되어야 할 위치가 겹치게 되기 때문

- z-index 특징

    - z-index : 요소가 겹쳤을 때 어떤 요소 순으로 위에 나타낼 지 결정
        
        => (덮어쓰기 등등)

    - 정수 값을 사용해 z축 순서를 지정
    - 더 큰 값을 가진 요소가 작은 값의 요소를 덮음

```python
<!-- position 예시 -->

<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<meta http-equiv="X-UA-Compatible" content="IE=edge" />
		<meta name="viewport" content="width=device-width, initial-scale=1.0" />
		<title>Document</title>
		<style>
			* {
				box-sizing: border-box;
			}

			body {
				height: 1500px;
			}
			/* container는 주로 여러 item을 묶어줄 때 활용(item 상위개념) */
			.container {
				position: relative;
				height: 300px;
				width: 300px;
				border: 1px solid black;
			}

			.box {
				height: 100px;
				width: 100px;
				border: 1px solid black;
			}

			.static {
				position: static;
				background-color: lightcoral;
			}

			/* normal flow 없애고 absolute 새로운 부모 찾기 => container 기준으로 */
			.absolute {
				position: absolute;
				background-color: lightgreen;
				top: 100px;
				left: 100px;
			}

			/* 본인 왼쪽 위의 점 기준 */
			.relative {
				position: relative;
				background-color: lightblue;
				top: 100px;
				left: 100px;
			}

			.fixed {
				position: fixed;
				background-color: gray;
				top: 0;
				right: 0;
			}
		</style>
	</head>
	<body>
		<div class="container">
			<div class="box static">static</div>
			<div class="box absolute">absolute</div>
			<div class="box relative">relative</div>
			<div class="box fixed">fixed</div>
		</div>
	</body>
</html>
```

```python
<!-- sticky 예시 -->

<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<meta http-equiv="X-UA-Compatible" content="IE=edge" />
		<meta name="viewport" content="width=device-width, initial-scale=1.0" />
		<title>Document</title>
		<style>
			body {
				height: 1500px;
			}

			/* 상단 메뉴 바 만들 때 활용 */
			.sticky {
				position: sticky;
				top: 0;
				background-color: lightblue;
				padding: 20px;
				border: 2px solid black;
			}
			.container {
				border: 3px solid red;
			}
		</style>
	</head>
	<body>
		<h1>Sticky positioning</h1>
		<div class="container">
			<div class="sticky">첫 번째 sticky</div>
			<div>
				<p>내용1</p>
				<p>내용2</p>
				<p>내용3</p>
			</div>
			<div class="sticky">두 번째 sticky</div>
			<div>
				<p>내용4</p>
				<p>내용5</p>
				<p>내용6</p>
			</div>
			<div class="sticky">세 번째 sticky</div>
			<div>
				<p>내용7</p>
				<p>내용8</p>
				<p>내용9</p>
			</div>
		</div>
	</body>
</html>
```

## CSS layout flexbox

- 요소를 행과 열 형태로 배치하는 *1차원* 레이아웃 방식
     
     <!-- display : flex or display inline-flex -->
     => **'공간 배열' & '정렬'**



![이미지](../이미지/flexbox.PNG)

- main axis : 주 축
     - flex item들이 배치되는 기본 축
     - main start ~> main end 방향으로 배치

- cross axis : 교차 축
     - main axis에 수직인 축
     - cross start ~ cross end 방향으로 배치
