## 11-1 CPU 스케줄링 개요

- CPU 스케줄링 : 운영체제가 프로세스들에게 공정하고 합리적으로 CPU 자원을 배분하는 것

#### 1. 프로세스 우선순위(priority)

- 가장 공정한 CPU 스케줄링? : 우선순위를 정해보자


1. 프로세스 우선순위

![이미지](../../이미지/프로세스우선순위.png)

- 입출력 작업이 많은 프로세스(=입출력 집중 프로세스)의 우선순위는 CPU 작업이 많은 프로세스(=CPU 집중 프로세스)의 우선순위보다 높다. 

- 우선순위는 PCB에 저장됨


2. 스케줄링 큐

![이미지](../../이미지/스케줄링%20큐.png)

- 프로세스들아 특정 자원을 이용하고 싶으면 "스케줄링 큐에서" 줄 서!

- 준비 큐와 대기 큐

    ![이미지](../../이미지/준비큐대기큐.png)

    - 같은 큐 내에서도 우선순위별로 처리
    
    - 대기큐 : 같은 장치를 요구한 프로세스들은 같은 큐에서 대기

    ![이미지](../../이미지/대기큐.png)


3. **선점형과 비선점형 스케줄링**

 ![이미지](../../이미지/선점형비선점형프로세스.png)

- A 프로세스가 실행중인데, B 프로세스가 정말 급할 때? 

    1. 선점형 스케줄링 
        
        - 현재 CPU를 사용 중인 프로세스로부터 CPU 자원을 **빼앗아** 다른 프로세스에 할당

        - 프로세스 B :  "비켜 ! 내가 쓸거야 !"(보통 정해진 시간만큼 CPU 쓰기)

        - 장점 : 어느 한 프로세스의 자원 독점을 막고 프로세스들에 골고루 자원을 배분할 수 있다. 

        - 단점 : 그만큼 문맥 교환 과정에서 오버헤드가 발생할 수 있다. 


    2. 비선점형 스케줄링
        
        - 현재 CPU를 사용 중인 프로세스의 작업이 끝날 때까지 프로세스 기다리기

        - 프로세스 A : "B 기다려 !"

        - 장점 : 선점형 스케줄링에 비해 문맥 교환에서 발생하는 오버헤드가 적다.

        - 단점 : 모든 프로세스가 골고루 자원을 이용하기 어렵다.
