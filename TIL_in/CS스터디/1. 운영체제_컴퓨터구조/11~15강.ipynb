## 11강 : 명령어 사이클과 인터럽트


#### 1. 명령어 사이클

1. 인출 & 실행

- 프로그램 속 명령어들은 일정한 주기가 반복되며 실행, 이 주기를 **명령어 사이클**이라고함

![이미지](../../이미지/인출실행.png)

- 인출 사이클 : 메모리에 저장된 명령어를 실행하려면, 가장 먼저 CPU로 갖고 와야함

- 실행 사이클 : 갖고 왔으면 실행해야 한다.

- 그런데 CPU로 명령어를 가지고 와도 바로 실행이 불가능한 경우도 있음(간접 주소 지정 방식)

![이미지](../../이미지/인출실행간접.png)

- 이러한 정해진 흐름을 끊어 버리는 것이 인터럽트임

![이미지](../../이미지/인터럽트.png)

2. 인터럽트

- 인터럽트(interrupt) : 방해하다, 중단시키다.

    - 예시 : 'CPU가 꼭 주목해야할 때', 'CPU가 얼른 처리해야 할 다른 작업이 생겼을 때' 발생!
    - 예시1 : "강대리, 이거 급한거니까 지금 하던 일 멈추고 이것부터 처리해주게"

    ![이미지](../../이미지/인터럽트2.png)

    1. 동기 인터럽트(예외) : CPU가 예기치 못한 상황을 접했을 때 발생

    ![이미지](../../이미지/동기인터럽트.png)

    2. 비동기 인터럽트(하드웨어 인터럽트) : 주로 입출력장치에 의해 발생

    ![이미지](../../이미지/비동기인터럽트.png)

    - 알림(세탁기 완료 알림, 전자레인지 조리 알림)과 같은 역할

    - **입출력 작업 도중에도 효율적으로 명령어를 처리**하기 위해 하드웨어 인터럽트 사용

        - 입출력장치는 CPU에 비해 느리다. 인터럽트가 없다면 CPU는 프린트 완료 여부를 확인하기 위해 주기적으로 확인해야한다. 하지만 인터럽트가 있다면 CPU는 입출력 작업동안 다른 일을 하고 있을 수 있다.(효율적임!)

3. 하드웨어 인터럽트 처리 순서

    1. 입출력장치는 CPU에 **인터럽트 요청 신호(알림)**를 보낸다
    2. CPU는 실행 사이클이 끝나고 명령어를 인출하기 전 항상 인터럽트 여부를 확인한다.
    3. CPU는 인터럽트 요청을 확인하고 **인터럽트 플래그**를 통해 현재 인터럽트를 받아들일 수 있는 지 여부를 확인한다.
    4. 인터럽트를 받아들일 수 있다면 CPU는 지금까지의 작업을 백업한다.(지금의 작업을 잠시 중단하기위해 스택 영역에 백업해둠)
    5. CPU는 **인터럽트 벡터**를 참조하여 **인터럽트 서비스 루틴**을 실행한다.
    6. 인터럽트 서비스 루틴 실행이 끝나면 ④에서 백업해 둔 작업을 복구하여 실행을 재개한다.


4. 하드웨어 인터럽트 용어 정리

    1. 인터럽트 요청 신호

    - CPU의 작업을 방해하는 인터럽트에 대한 요청
    - 입출력장치가 CPU에게 지금 끼어들어 되나요? 요청

    2. 인터럽트 플래그

    - 인터럽트 요청 신호를 받아들일지 무시할지를 결정하는 비트

    ![이미지](../../이미지/하드웨어%20인터럽트.png)

    3. 인터럽트 서비스 루틴

    - CPU가 인터럽트를 받아들이기로 했다면 인터럽트 서비스 루틴 실행(인터럽트를 처리하는 프로그램)

    - 인터럽트 서비스 루틴이란, 인터럽트가 발생했을 때 인터럽트를 어떻게 처리하기 위한 프로그램(루틴으로 정리되어 메모리에 저장되어 있음)

    ![이미지](../../이미지/인터럽트%20서비스%20루틴.png)

    4. 인터럽트 벡터 : 각각의 인터럽트를 구분하기 위한 정보

    - 인터럽트 서비스 루틴의 시작 주소를 포함하는 인터럽트 서비스 루틴의 식별 정보

    ![이미지](../../이미지/인터럽트벡터.png)

## 12강 : 빠른 CPU를 위한 설계 기법

- 컴퓨터 부품들은 '클럭 신호'에 맞춰서 일사분란하게 움직인다.
- CPU는 '명령어 사이클'이라는 정해진 흐름에 맞춰 명령어들을 실행한다.

- 빠른 CPU를 설계하기 위해서는..
    
    - 클럭 신호의 주기를 빠르게 반복하면 되지 않을까? : 일반적으로는 Yes

    - but 필요 이상으로 클럭을 높이면 **발열**이 심해짐

#### 1. 클럭(clock)

- 클럭속도 : Hz(헤르츠) 단위로 측정
    
    - 클럭이 '똑-딱-'하고 1초에 한 번 반복되면 1Hz

    - 클럭이 1초에 100번 반복되면 100Hz

- 클럭 속도를 늘리는 방법 이외에는?

    1. **코어 수**를 늘리는 방법(듀얼코어, 멀티 코어 등)

    2. **스레드 수**를 늘리는 방법(멀티 스레드 등등)


#### 2. 코어 & 멀티 코어

1. 코어(Core)란?

    ![이미지](../../이미지/코어.png)

    - CPU내에서 '명령어를 실행하는 부품'을 코어라는 용어로 사용되며, 여러 개 있을 수 있음

    ![이미지](../../이미지/코어1.png)

    - 코어를 2개, 3개, 100개 늘리면 연산속도도 그에 비례하여 빨라지나요?
    
    - 꼭 코어 수에 비례하여 증가하지는 않습니다.(사공이 많으면 배가 산으로 간다)

    - 중요한 건, 명령어들을 코어들에게 얼마나 적절하게 분배하느냐가 중요함


#### 3. 스레드 & 멀티 스레드

![이미지](../../이미지/스레드.png)

1. 스레드 : 실행 흐름의 단위


2. 하드웨어적 스레드(논리 프로세서)

- 하나의 코어가 동시에 처리하는 명령어 단위

    - 1코어 1스레드 CPU

    ![이미지](../../이미지/1코어1스레드.png)

    - 2코어 4스레드 CPU(멀티스레드 CPU) == 각 코어가 동시에 명령어 2개 수행 가능, 즉 한번에 4개 명령어 수행 가능

    ![이미지](../../이미지/2코어4스레드.png)

- **레지스터** : 멀티스레드 프로세서를 설계하는데 가장 큰 핵심!

    - 한 코어에 레지스터 세트가 여러 개라면, 여러 명령어를 빠르게 처리할 수 있으므로 멀티스레드 CPU가 가능해진다.


3. 소프트웨어적 스레드

- 하나의 프로그램에서 독립적으로 실행되는 단위

    ![이미지](../../이미지/멀티스레드.png)

- 1코어 1스레드 CPU도 여러 소프트웨어적 스레드를 만들 수 있다. 






## 13강 : 명령어 병렬 처리 기법

- CPU가 **시간을 알뜰하게 쓰며** 저 명령어들을 처리할 방법이 있을까?

#### 1. **명령어 파이프라인**

- 명령어가 처리되는 과정을 비슷한 시간 간격으로 나누면?

    - 명령어 인출 -> 명령어 해석 -> 명령어 실행 -> 결과 저장

    - 각 기간동안 겹쳐서 실행함으로써 성능 향상(컨베이어 벨트)

    ![이미지](../../이미지/명령어%20파이프라인.png)

#### 2. 파이프라인 위험

- 명령어 파이프라인이 성능 향상에 실패하는 경우

    ![이미지](../../이미지/파이프라인%20위험.png)

    1. 데이터 위험

    - **명령어 간의 의존성**에 의해 야기

    - 모든 명령어를 동시에 처리할 수는 없다.(이전 명령어를 끝까지 실행해야만 비로소 실행할 수 있는 경우)

    ![이미지](../../이미지/데이터%20위험.png)

    2. 제어 위험

    - 프로그램 카운터의 갑작스러운 변화(JUMP 명령어 등)

    ![이미지](../../이미지/제어위험.png)

    3. 구조 위험

    - 서로 다른 명령어가 같은 CPU 부품(ALU, 레지스터)를 쓰려고 할 때


#### 3. 슈퍼스칼라

- CPU 내부에 여러 개의 명령어 파이프라인 포함한 구조

- 이론적으로는 파이프라인 개수에 비례하여 처리 속도 증가하지만, **파이프라인 위험도**의 증가로 인해 파이프라인 개수에 비례하여 처리 속도가 증가하진 않음


#### 4. 비순차적 명령어 처리

- 파이프라인의 중단을 방지하기 위해 명령어를 순차적으로 처리하지 않는 명령어 병렬 처리 기법

- 즉, 명령어들 간의 합법적인 새치기

## 14강 : 명령어 집합 구조, CISC와 RISC

- 명령어가 어떻게 생겨야 명령어 파이프라이닝에 유리할까요?

#### 1. 명령어 집합(구조)

- 명령어 집합(구조) : CPU가 이해할 수 있는 명령어들의 모음, CPU의 언어인 셈

- CPU는 명령어를 실행한다. 하지만 인텔 CPU 컴퓨터에서 만든 실행파일을 그대로 아이폰에 옮겨 특별한 설정없이 바로 실행하면 실행이 안된다. 즉, 각 제조사의 CPU는 조금씩 다르고 그에 대한 명령어 해석 방식, 레지스터의 종류와 개수, 파이프라이닝 용이성 등등 에 대해 조금씩 다르다.

#### 2. CISC & RISC(명령어 집합의 두 축)

1. CISC

- Complex Instruction Set Computer(**복잡한** 명령어 집합을 활용하는 컴퓨터)

    - 명령어의 형태와 크기가 다양한 **가변 길이 명령어**를 활용

    - 상대적으로 적은 수의 명령어로도 프로그램을 실행할 수 있다.

- 이러한 점은 메모리를 최대한 아끼며 개발해야 했던 시절에 인기가 높았으나, **명령어 파이프라이닝이 불리하다**는 치명적인 단점이 존재함.

    - 명령어가 워낙 복잡하고 다양한 기능을 제공하는 탓에 **명령어 크기와 실행되기 까지의 시간이 일정하지 않음**

    - 복잡한 명령어로 인해 **명령어 하나를 실행하는 데에 여러 클럭 주기** 필요 => 명령어 파이프라이닝에 불리

2. RISC

- Reduced Instruction Set Computer(명령어의 종류가 적고, 짧고 규격화된 명령어 사용하는 컴퓨터)

    - 명령어 파이프라이닝에 유리함

    - 메모리 접근 최소화(load, store), 레지스터 십분 활용(따라서 레지스터 종류 많음)

    - 다만 명령어 종류가 CISC보다 적기에 더 많은 명령어로 프로그램을 동작시킴

## 15강 : RAM의 특성과 종류(메모리)

- 주기억장치의 종류에는 크게 RAM과 ROM 두 가지가 있고, '메모리'라는 용어는 그 중 RAM을 지칭하는 경우가 많다.

- RAM이 크면 정확히 뭐가 좋을까?

#### 1. RAM의 특징

![이미지](../../이미지/램과%20보조기억장치.png)

#### 2. RAM의 용량과 성능

- RAM의 용량과 성능이 좋을수록, 많은 프로그램들을 동시에 실행하는 데 유리함

#### 3. RAM의 종류 : DRAM, SRAM, SDRAM, DDR SDRAM

1. DRAM(dynamic RAM)

- 저장된 데이터가 **동적으로 사라지는** RAM

- 데이터 소멸을 막기 위해 주기적으로 재활성화(refresh) 해야함

- 일반적으로 메모리로 사용되는 RAM(상대적으로 소비전력이 낮고 저렴하고 집적도가 높아 대용량으로 설계하기 용이하기 때문에)

2. SRAM(static RAM)

- 저장된 데이터가 **정적인(사라지지 않는)** RAM

- DRAM 보다 일반적으로 더 빠름

- 일반적으로 캐시 메모리에서 사용되는 RAM(상대적으로 소비전력이 높고 가격이 높고 집적도가 낮아, "대용량으로 설계할 필요는 없으나 빠르게 동작해야하는 장치"에 사용)

3. DRAM vs SRAM

![이미지](../../이미지/디램에스램.png)


4. SDRAM(synchronous DRAM)

- 동기화된 DRAM(발전된 형태의 디램)

- 클럭 신호와 동기화된 DRAM


5. DDR SDRAM(double data rate SDRAM)

- 발전된 형태의 SDRAM

- 최근 가장 대중적으로 사용하는 RAM

- 대역폭(데이터를 주고받는 길의 너비)을 넓혀 속도를 빠르게 만든 SDRAM


6. DDR2,3,4 SDRAM

- DDR SDRAM보다 대역폭이 2배, 4배, 8배 넓은 RAM
