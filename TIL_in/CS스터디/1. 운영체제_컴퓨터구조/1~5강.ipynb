## 4강 : 0과 1로 숫자를 표현하는 방법(데이터)

- 컴퓨터는 0과 1밖에 모르는 바보..

- 근데 3+4 = 7 이라는 것을 어떻게 알지?

#### 1. 정보 단위

1. 비트(bit), 바이트(byte)

    - 비트(bit) : 0과 1을 표현하는 가장 작은 정보 단위(2가지 정보   표현 가능)

    - n비트로 2^n가지의 정보 표현 가능(프로그램은 수많은 비트로   이루어져 있음)

    - 평소 비트보다 더 큰 단위를 사용 : 바이트, 킬로바이트,   메가바이트, 기가바이트, 테라바이트

    ![이미지](../../이미지/비트_바이트.png)

2. 워드(word)

    - CPU가 한 번에 처리 할 수 있는 정보의 크기 단위

    - 하프 워드(half) : 워드의 절반 크기

    - 풀 워드(full) : 워드 크기

    - 더블 워드(double) : 워드의 두 배 크기


#### 2. 이진법(0,1로 숫자 표현하기)

1. 양수 표현하기(8)
    
    - 1000(2)
    - ob1000

2. **음수 표현하기(-) : 2의 보수**

    - 2의 보수 : 어떤 수를 그보다 큰 2^n에서 뺀 값

        - 예시 : 11(2)을 음수로 표현 
        
            - 11(2)보다 큰 2^n = 100(2)

            - 100(2) - 11(2) = 01(2)

            - 11(2)을 음수로 표현한 값 01(2)

        - **모든 0과 1을 뒤집고, 1 더한 값** 

            - 11(2)을 음수로 표현

            - 11 -> 00 -> 01

            - 11(2)을 음수로 표현한 값 01(2)

            - 플래그

## 5강 : 0과 1로 문자를 표현하는 방법(데이터)

- 0과 1밖에 모르는 컴퓨터가 어떻게 문자를 이해하고 표현할 수 있지?


#### 1. 문자 집합과 인코딩

![이미지](../../이미지/문자집합.png)

- 문자 집합(character set) : 컴퓨터가 이해할 수 있는 문자의 모음

- 인코딩(encoding) : 코드화하는 과정, 문자를 0과 1로 이루어진 문자 코드로 변환하는 과정 

- 디코딩(decoding) : 코드를 해석하는 과정, 0과 1로 표현된 문자 코드로 문자로 변환하는 과정

#### 2. 문자 집합들 

1. 아스키 코드(ASCII)

    - 초창기 문자 집합 중 하나(알파벳, 아라비아 숫자, 일부 특수문자 등)

    - 7비트로 하나의 문자 표현(8비트 중 1비트는 오류 검출을 위해 사용되는 패리티 비트)


2. EUC-KR(국가별, 언어별 인코딩)

    - 문자집합 기반의 한글 인코딩 방식(완성형 인코딩)

    - 글자 하나 하나에 2바이트 크기의 코드 부여(2바이트 == 16비트)


3. 유니코드 문자 집합과 UTF-8

    1. 유니코드

        - 통일된 문자 집합 & 인코딩 방식(utf-8, utf-16, utf-32...)

        - 한글, 영어, 특수문자, 이모티콘

        - 현대 문자 표현에 있어 매우 중요한 위치

    2. UTF-8 인코딩

        - UTF(unicode transformation format) == 유니코드 인코딩 방법

        - 가변 길이 인코딩 : 인코딩 결과가 1 ~ 4 바이트 

        - 인코딩 결과가 몇 바이트가 될지는 유니코드에 부여된 값에 따라 다름
