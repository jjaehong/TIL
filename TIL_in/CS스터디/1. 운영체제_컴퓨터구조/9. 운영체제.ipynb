## 운영체제

#### 9-1 운영체제를 알아야 하는 이유

- 모든 프로그램은 실행을 위해 **자원**을 필요로 한다.

- 자원/시스템 자원 : 프로그램 실행에 있어 마땅히 필요한 요소(컴퓨터 4가지 핵심부품 포함)

1. 운영체제 개념(Windows, macOS, Linux, iOS, android)

  - 운영체제란 실행할 프로그램에 필요한 **자원을 할당하고** 프로그램이 **올바르게 실행되도록** 돕는 특별한 프로그램

  1. 운영체제 메모리 관리

  ![이미지](../../이미지/운영체제.png)

  - 운영체제는 메모리의 커널 영역에 적재되는 프로그램임

  - 운영체제는 프로그램 적재 후 실행, 종료했다면 메모리에서 삭제 후 비워둠


  2. 운영체제의 CPU 관리 

  ![이미지](../../이미지/운영체제1.png)
  
  - 프로그램 멀티태스킹에서 우선순위 관리(교통 정리)


  3. 운영체제의 입출력장치 관리

  ![이미지](../../이미지/운영체제2.png)


  4. 정리

  ![이미지](../../이미지/운영체제3.png)

  - 운영체제란 실행하는 응용프로그램과 하드웨어 사이에 위치하여 둘 사이를 연결시켜주는 프로그램이다. 
  
  - 운영체제덕분에 개발자는 하드웨어를 조작하는 코드를 직접 작성할 필요가 없다.

  - 운영체제는 사용자를 위한 프로그램이 아니라 **프로그램을 위한 프로그램이다.** 그렇기에 프로그램을 만드는 개발자는 운영체제를 알아야한다. 운영체제에 어떻게 도움을 받을 수 있을지 생각하고 오류 메시지에 대해 깊은 이해를 통해 문제해결능력을 기른다. 





#### 9-2 운영체제의 큰 그림



1. 커널(운영체제의 심장)

- 운영체제의 핵심 서비스를 담당하는 부분을 커널(kernel)이라고 한다.

- 운영체제에는 속하는데 커널에는 속하지 않는 기능? : UI(유저 인터페이스)

    - UI는 사용자와 컴퓨터 간의 통로(마우스, 터치로 조작)일 뿐 운영체제의 핵심 기능(커널)은 아님


2. 운영체제의 핵심 서비스 종류

    1. 프로세스 관리

    - 프로세스 == 실행 중인 프로그램, 수많은 프로세스들이 동시에 실행

    - 동시다발적으로 생성/실행/삭제되는 다양한 프로세스를 **일목요연하게 관리**

        - 프로세스와 스레드

        - 프로세스 동기화

        - 교착상태 해결



    2. 자원 접근 및 할당

    - CPU 접근 및 할당

        - CPU 스케줄링 : 어떤 프로세스를 먼저, 얼마나 오래 실행할까?

    - 메모리 접근 및 할당

        - 페이징, 스와핑 등등 

    - 입출력장치 접근 및 할당

        - 인터럽트 서비스 루틴


    3. 파일 시스템 관리


3. 시스템 콜(호출)과 이중 모드

- 사용자가 실행하는 프로그램은 자원에 직접 접근할 수 있을까? => NO, 자원에 직접 접근은 위험하다. 

- 운영체제는 응용프로그램이 자원에 접근하려 할 때, 오직 **자신을 통해서만 접근하도록** 하여 자원을 보호

    1. 이중 모드

    - CPU가 명령어를 실행하는 모드를 크게 **사용자 모드**와 **커널 모드**로 구분하는 방식

        1. 사용자 모드

            - 운영체제 서비스를 제공받을 수 없는 실행 모드

            - 커널 영역의 코드를 실행할 수 없는 실행 모드

            - 자원 접근 불가

        2. 커널 모드

            - 운영체제의 서비스를 제공받을 수 있는 실행 모드

            - 자원 접근을 비롯한 모든 명령어 실행 가능

        ![이미지](../../이미지/이중모드.png)

    2. 시스템 호출

    - 커널 모드로 전환하여 실행하기 위해 호출

    - 일종의 **소프트웨어 인터럽트**

        ![이미지](../../이미지/시스템%20호출.png)
