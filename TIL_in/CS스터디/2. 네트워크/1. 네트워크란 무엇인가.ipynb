## 1. 네트워크란 무엇인가?


#### 1. 네트워크란?

![이미지](../../이미지/네트워크.PNG)

- **노드**들이 데이터를 공유할 수 있게 하는 디지털 전기**통신망**의 하나이다.
- 즉, 분산되어 있는 컴퓨터를 통신망으로 연결한 것을 의미한다.
- 네트워크에서 여러 장치들은 노드 간 연결을 사용하여 서로에게 데이터를 교환한다. 


#### 2. 인터넷이란?

- 문서, 그림, 영상과 같은 여러가지 데이터를 공유하도록 구성된 세상에서 가장 큰 전세계를 연결하는 네트워크


#### 3. 네트워크의 분류


1. 크기에 따른 분류

    ![이미지](../../이미지/wan_lan.PNG)

    - LAN(local area network) : LAN은 가까운 지역을 하나로 묶은 네트워크, 예시 : 스타크래프트 LAN서버.

    - WAN(wide area network) : WAN은 멀리 있는 지역을 한데 묶은 네트워크, 여러 개의 LAN을 서로 연결시켜둠.


2. 연결 형태에 따른 분류

    - Star형 : 중앙 장비에 모든 노드가 연결된 형태의 네트워크(**중앙화**)

    - Mesh형 : 여러 노드들이 서로 그물처럼 연결된 형태의 네트워크(**탈중앙화**)
    
    - 혼합형 : 실제 인터넷은 여러 형태를 혼합한 형태

    ![이미지](../../이미지/혼합형.png)



#### 4. 네트워크 통신방식

- 네트워크에서 데이터는 어떻게 주고 받는가?


1. 유니 캐스트

    - 특정 대상이랑만 1:1로 통신하는 유니 캐스트

2. 멀티 캐스트

    - 특정 다수와 1:N으로 통신하는 멀티 캐스트

3. 브로드 캐스트

    - 네트워크에 있는 모든 대상과 통신하는 브로드 캐스트


#### 5. 네트워크 프로토콜

1. 프로토콜이란? 

    - 프로토콜은 일종의 약속, 양식
    - 네트워크에서 노드와 노드가 통신할 때 **어떤 노드가 어느 노드에게 어떤 데이터를 어떻게 보내는지** 작성하기 위한 양식
    - 프로토콜은 프로토콜만의 양식이 있기에 양식에 맞춰서 보내줘야한다. 

2. 여러가지 프로토콜

    ![이미지](../../이미지/프로토콜%20종류1.PNG)

    - 가까운 곳, 멀리 있는 곳, 여러가지 프로그램 연락 등 에서 통신할 때 데이터를 무엇으로 주고받아`?(양식)
