## 2. 네트워크 모델

=> 네트워크의 기준

#### 1. 네트워크 계층 모델

1. TCP/IP 모델 

    - 인터넷에서 컴퓨터들이 서로 정보를 주고받는데 쓰이는 프로토콜의 모음

    - 네트워크 인터페이스(1계층)
    - 네트워크(2계층)
    - 전송(3계층)
    - 응용(4계층)

2. OSI 7계층

    - 1984년 이후 네트워크 표준으로 지정한 모델

    ![이미지](../../이미지/OSI7.PNG)


3. TCP/IP vs OSI 7 계층

    1. 공통점
    
        - 계층적 네트워크 모델
        - 계층간 역할 정의

    2. 차이점

        - 계층의 수 차이
        - OSI는 역할 기반(논리 기반), TCP/IP는 프로토콜 기반(실무 기반)
        - OSI는 통신 전반에 대한 표준
        - TCP/IP는 데이터 전송기술 특화

#### 2. 패킷 : 네트워크를 통해 전달되는 데이터


1. 패킷 개념
    
    - 네트워크 상에서 전달되는 데이터를 통칭하는 말

    - 네트워크에서 전달하는 데이터의 형식화된 **블록**이다

    - 제어 정보(누가 누구에게, 어떤 데이터를 어떻게 보내는지)와 사용자 데이터로 이루어지며, 사용자 데이터는 payload라고도 한다.
