## 2. 네트워크 모델

=> 네트워크의 기준

#### 1. 네트워크 계층 모델 종류

1. TCP/IP 모델 

    - 인터넷에서 컴퓨터들이 서로 정보를 주고받는데 쓰이는 프로토콜의 모음

    - 네트워크 인터페이스(1계층)
    - 네트워크(2계층)
    - 전송(3계층)
    - 응용(4계층)

2. OSI 7계층

    - 1984년 이후 네트워크 표준으로 지정한 모델

    ![이미지](../../이미지/OSI7.PNG)


3. TCP/IP vs OSI 7 계층

    1. 공통점
    
        - 계층적 네트워크 모델 : 내트워크를 계층으로 나눔
        - 계층간 역할 정의

    2. 차이점

        - 계층의 수 차이
        - OSI는 역할 기반(논리 기반), TCP/IP는 프로토콜 기반(실무 기반)으로 분리
        - OSI는 통신 전반에 대한 표준(면접 대비)
        - TCP/IP는 데이터 전송기술 특화

#### 2. 패킷 : 네트워크를 통해 전달되는 데이터


1. 패킷 개념
    
    - 네트워크 상에서 **전달되는 데이터**를 통칭하는 말이며, **데이터의 형식화된 블록**이다.

    - **제어 정보(누가 누구에게, 어떤 데이터를 어떻게 보내는지)와 사용자 데이터(무엇을 요청하고 응답해주는지)로 이루어지며**, 사용자 데이터는 payload라고도 한다.

2. 패킷 구성

    - 여러 프로토콜들로 캡슐화된 패킷
    
    ![이미지](../../이미지/페이로드.PNG)

    ![이미지](../../이미지/패킷1.PNG)
    
    - HTTP 프로토콜을 payload로 해서 TCP라는 헤더를 붙여서 패킷을 만듬
    - 다시 캡슐화된 패킷을 payload로 해서 IPv4를 헤더로 붙여서 패킷을 만듬
    - 다시 캡슐화된 패킷을 payload로 해서 Ethernet을 헤더로 붙여서 패킷으로 만듬
    - 따라서 패킷의 형태는 정해져 있지 않고 계속 변경될 수 있음

3. 캡슐화 & 디캡슐화

    1. 캡슐화
        
        - 여러 프로토콜을 이용해서 최종적으로 **보낼 때** 패킷을 만드는 과정
        - **상위 계층에서부터 하위 계층으로 내려가면서 프로토콜을 붙임**
     
        ![이미지](../../이미지/패킷%20캡슐화1.PNG)

    2. 디캡슐화 
    
        - 패킷을 **받았을 때** 프로토콜들을 하나씩 확인하면서 데이터를 확인하는 과정
        - 하위 프로토콜에서부터 상위 프로토콜로 하나씩 확인
        
        ![이미지](../../이미지/디캡슐화1.PNG)

4. PDU(Protocol Data Unit)

    - 계층별 패킷의 이름
    - segment : 4계층까지 합쳤을 때 PDU
    - 패킷 : 3계층까지 합쳤을 때 PDU
    - 프레임 : 2계층까지 합쳤을 때 PDU
