#### 2. 관계형 데이터베이스(RDB)

- 데이터베이스 역할 : 데이터를 **저장(구조적 저장)**하고 **조작(CRUD)**

1. 개념
    
    - 데이터 간에 **관계**가 있는 데이터 항목들의 모음

    - 테이블, 행, 열의 정보를 구조화하는 방식

    - **서로 관련된 데이터 포인터를 저장**하고 이에 대한 **액세스**를 제공(보통 id로 연결)

    ![이미지](../이미지/관계형db1.PNG)

    - 이 관계로 인해 두 테이블을 사용하여 데이터를 다양한 형식으로 조회할 수 있음

        - 특정 날짜에 구매한 모든 고객 조회

        - 지난 달에 배송일이 지연된 고객 조회 등


    ![이미지](../이미지/관계형db2.PNG)

    - **Primary Key(기본 키)** : 관계형 데이터베이스에서 **레코드의 식별자(각 레코드 고유한 값)**로 활용
    
    ![이미지](../이미지/관계형db3.PNG)

    - **Foreign Key(외래 키)** : 테이블의 필드 중 다른 테이블의 레코드를 식별할 수 있는 키

    - 다른 테이블의 기본 키를 참조

    - 각 레코드에서 서로 다른 테이블 간의 **관계를 만드는 데** 사용

    - 직접적인 연결이 되어있는 건 아니지만 외래키를 통해서 주문테이블이 고객 테이블을 참조하고 있다!

    ![이미지](../이미지/rdb키워드1.PNG)

#### 5. DQL

- data query language

1. SELECT 문 : 테이블에서 데이터를 **조회**

    - 활용 1 : 이 필드를 선택한다(SELECT) 이 테이블로부터(FROM);
    
    - 활용 3 : SELECT * FROM employees; : employees테이블의 모든 필드 데이터 조회

    - 활용 5 : SELECT Name, Milliseconds / 60000 AS '재생 시간(분)' FROM tracks;

        - 테이블 tracks에서 Name 필드, 재생시간(분) 필드로 이름바꿔 모든 데이터 조회

#### 6. DDL

- data definition language : 데이터의 기본 구조 및 형식 변경

    - ex : CREATEM, DROP, ALTER

1. CREATE TABLE : 테이블 생성

![이미지](../이미지/createtable.PNG)

![이미지](../이미지/createtable1.PNG)

- 데이터 타입(dtype)

    - NULL : 아무런 값도 포함하지 않음
    - INTEGER : 정수
    - REAL : 부동 소수점
    - TEXT : 문자열
    - BLOB : 이미지, 동영상, 문서 등의 바이너리 데이터

- 제약 조건(constraints) : 테이블 필드에 적용되는 규칙 또는 제한사항

    - PRIMARY KEY : 해당 필드를 기본 키로 지정, INTEGER 타입에만 적용되며, INT, BIGINT 등의 정수 유형은 적용되지 않음
        
    - NOT NULL : 해당 필드에 NULL 값을 허용하지 않도록 지정

    - FOREIGN KEY : 다른 테이블과의 외래 키 관계를 정의

- AUTOINCREMENT keyword : 자동으로 고유한 정수 값을 생성하고 할당하는 필드 속성(필드 자동 증가)
