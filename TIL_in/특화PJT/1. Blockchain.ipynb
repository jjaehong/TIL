#### 1. 프로젝트 개요


1. 프로젝트 개요

- 블록체인을 활용한 **P2P 전자상거래** 시스템 구현

- 우리가 구현할 전자상거래 시스템은 사용자가 거래 물품을 플랫폼 상에 등록하여 판매자가 되기도 하고, 거래 물품의 구매자가 되기도 한다.

- 거래 물품을 등록하거나 구매할 때, 암호화폐나 블록체인 상의 자산인 토큰으로 거래하고 스마트 컨트랙트를 통해 판매-구매 프로세스를 진행합니다. 

- **이커머스 플랫폼의 구매, 지불-결제 기능을 블록체인을 활용하여 구현**하는 것이 핵심


2. 활용 사례

    1. 자전거 도난 방지 시스템

        - 전기 자전거의 고유 번호를 기반으로 자전거의 정확환 위치, 상태 등을 기록, 추적하고 도난 발생 시 신고, 추적에 드는 사회적 비용을 줄이는데 효용을 보임

    2. 트레이드렌즈

        - 무역, 물류의 생애주기를 블록체인으로 관리하고 종이 계약서를 디지털화한 사례로 유명함, 컨테이너를 실시간으로 상태 추적 가능

    3. 에버렛저

        - 귀중품을 위한 공급망 플랫폼
        - 블록체인 상에 인증서를 발행하여 관리하고 거래 및 소유 내역도 추가로 저장할 수 있도록 한 사례

    4. 정리

        - 이처럼 거래, 유통 시장의 신뢰도 제고와 업무 효율을 위해 블록체인이 도입되어 이제는 상용화 수준에 이름
        - 유통시장에서 블록체인 도입의 장점은 비교적 적은 비용으로도 신뢰가능한 지불 체계 구현이 가능해졌다는 뜻
        - 중개자 개입 없이도 신뢰가능한 거래가 가능하기에 제3자 혹은 플랫폼에 지불해야 했던 **중개 수수료가 낮아짐**

    5. 에스크로(Escrow)

        - 에스크로란 상거래 시, 판매자와 구매자 사이에 신뢰할 수 있는 제 3자를 통해 금전 혹은 물품을 거래하도록 하는 것을 말한다.
        - 블록체인과 스마트 컨트랙트를 사용하면 중개자 없이도 동일한 에스크로 기능을 구현할 수 있음


3. 유의 사항

    - 블록체인은 탈중앙화다.
    - '실제 서비스로 제공하였을 때에는 개발 프로젝트와 어떤 부분이 달라져야하는지?'
    - '어떤 부분은 중앙화하고 어떤 부분은 탈중앙화해야하는지?'
    - 블록체인을 도입하기 위해 고민을 하다보면 비즈니스 생태계를 둘러싼 현존하는 중앙화 시스템을 탈중앙화 한다는 것이 얼마나 도전적인 것이며, 도전적인 만큼 그 실현의 결과물이 기존 산업 구조를 뒤바꿀 수 있는 파격적 수단이 될 수 있을지에 대한 시각을 갖는 좋은 기회가 될 수 있을 것이라 생각합니다.

#### 2. 필수 지식 학습

1. 이더리움

    - 이더리움은 블록체인 기술을 기반으로 하는 분산 컴퓨팅 플랫폼이다.
    - 스마트 컨트랙트 기능
    - 암호 해시함수
        - 이더리움은 ASIC 저항성을 향상하고 채굴의 중앙화를 해소하기 위해 Ethash 라는 KECCACK 기반의 해시 알고리즘을 개발하여 사용함
    -  MPT(modify merkle patricia trie)
        - 이더리움에서는 MPT를 사용하여 상태 정보를 key-value 형식으로 저장하며 관리함
    - 암호화폐 관리 기술에 관한 차이
        - "사용자 계정"에 기반하여 암호화폐를 관리함

2. Geth(Go-Ethereum)

    - Geth는 이더리움 클라이언트
    - Geth는 이더리움에서 제공하는 다양한 API를 사용해볼 수 있으며 메인넷, 테스테넷 등과의 연동을 쉽게 할 수 있도록 편의성을 제공

3. Web3.js

    - 이더리움 생태계에 다양한 요청을 처리할 수 있게 하는 JS API 라이브러리
    - Geth와 Web3.js는 JSON RPC를 지원하고 있음
    - Web3.js는 JS 인터페이스만으로도 이더리움 노드들과 상호작용을 가능케 하므로 DApp이나 이더리움 기반 서비스 개발에 상당한 편의를 제공하고 있음.

4. Solidity

    - Solidity는 이더리움에서 개발한 스마트 컨트랙트 개발 언어
    - Solidity 스마트 컨트랙트는 컴파일러를 통해 바이트 코드로 변환되어 이더리움 가상 머신 위에서 동작하게 됨.
    - 이더리움의 주소를 나타내는 address 자료형 존재
    - 레퍼런스 타입의 구조체, 배열, 열거형, 매핑형 사용 가능
    - 매핑형은 해시 테이블과 같으며 Key에는 기본형을, value에는 사용자 지정 타입까지 지정할 수 있음

5. VirtualBox

    - 가상화를 위한 소프트웨어로, 다양한 운영체제에서 사용할 수 있음


6. Vagrant

    - 가상 머신 프로비저닝 도구
    - 간단한 스크립트를 작성하여 VirtualBox, Hyper-V 등 다양한 가상 머신을 쉽게 생성, 수정, 삭제, 관리할 수 있음.
    - Vagrant를 사용하면 설정 파일 하나로 모든 가상머신을 통합 관리할 수 있어 많은 편리함 제공

7. Remix IDE

    - 별도 설치 없이 웹 브라우저 상에서 이용가능한 이더리움 스마트 컨트랙트 개발 도구

8. MetaMask

    - 이더리움 지갑 프로그램의 일종
