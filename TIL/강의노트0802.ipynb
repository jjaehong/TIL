## 2차원 배열


#### 2차원 배열 선언

- 2차원 이상의 다차원 list는 차원에 따라 index를 선언


- 다차원 list
    - arr1 = [0]
    - arr2 = [0][0]
    - arr3 = [0][0][0]
    
    - arr2 = [
        [0,1,2,3]
        [4,5,6,7]
    ]


#### 2차원 배열 입력받기

N = int(input())

arr2 = [list(map(int, input())).split for _ in range(N)] 

    - list() => 한 줄 입력받기 
    - [list() ,for _ in range(N)] => 한 줄을 N번 입력 받아서 2차원 배열로 만든다

#### 2차원 배열의 접근

배열 순회 : n*m 배열의 모든 원소를 빠짐없이 조사하는 방법

1. 행 우선순회
- i행의 좌표, j열의 좌표

```python
for i in range(n):
    for j in range(m):
        f(Array[i][j]) # 필요한 연산 수행
```


2. 열 우선순회
- i행의 좌표, j열의 좌표

```python
for j in range(m):
    # max_v = 0
    # s = 0 : 초기값
    for i in range(n):
        f(Array[i][j]) 

        # if max_v < s 등등

```


3. 지그재그 순회
- i행의 좌표, j열의 좌표

```python
for i in range(n):
    for j in range(m):
        f(Array[i][j + (m-1-2*j)*(i%2)]) 
        # 증가하는 방향 : A[i][j], 
        # 감소하는 방향 : A[i][m-1-j]
        # i % 2 == 0 : j  //// i는 0부터..
        # i % 2 == 1 : (m-1) - j
        # 필요한 연산 수행
```



````python
# 2차원 배열 응용(행 총합의 최댓값)

N = 2
M = 4 
arr = [[0,1,2,3],
       [4,5,6,7]]

max_v = 0
for i in range(N):
    row_total = 0 # 초기값
    for j in range(M):
        row_total += arr[i][j]
    if max_v < row_total:
        max_v = row_total

print(max_v)

````

````python
import sys
sys.stdin = open("in.txt", "r")


T = int(input())
for tc in range(1,T+1):
    N = int(input())
    matrix = [list(map(int, input().split())) for _ in range(N)] # 2차원 배열 입력 받기 # list() 한줄 입력받기 # [ ~ ,for _ in range(N)] => N번 입력 받아서 2차원 배열로 만든다

    # 상하좌우(델타) = (행의좌표, 열의좌표) -> (i,j) (r,c)

    # idx[]상 하 좌 우
    dr = [-1, 1, 0, 0] # 행 번호의 변화량
    dc = [0, 0, -1, 1] # 열 번호의 변화량
    # 상(-1,0) / 하(1,0) / 좌(0,-1) / 우(0,1) 만큼 변한다.


    # 행 우선
    diff_sum = 0
    for r in range(N): # 가로길이 = 세로길이 / 행 = 열 = N
        for c in range(N):
            #print(f'현재위치 : ({r},{c}), 값 : {matrix[r][c]}') # 현재 위치가 r행 c열 (r,c) # matrix[r][c] => 현재 위치에 저장된 값

            now_diff_sum = 0 # 현재 위치에서 상하좌우 차이 절대값의 합

            # 상하좌우 좌표 탐색 
            for d in range(4): # d는 델타 배열의 인덱스[0,1,2,3], d = 0이면 상...
                # 현재 위치는 (r,c) 상하좌우로 이동 후 위치를 (nr,nc)
                nr = r + dr[d] # 다음 행 좌표 = 현재 행 좌표 + 이동방향이 d일 때 변화량
                nc = c + dc[d] # 다음 열 좌표 = 현재 열 좌표 + 이동방향이 d일 때 변화량
                # print(f' 다음 행렬 좌표는 ({nr}, {nc})')



                # 내가 계산한 위치가 유효한 인덱스인지?(배열의 범위 안에 있는지) 검사
                if 0 <= nr < N and 0 <= nc < N:
                    # 차이 = 현재 위치에 있는 값 - 새로 움직인 위치에 있는 값
                    diff = matrix[r][c] - matrix[nr][nc]

                    # 절대값 abs() 사용하지 않고 구하기
                    diff = -diff if diff < 0 else diff
                    now_diff_sum += diff

                # 상하좌우 탐색이 끝나면 현재 위치 절대값 차이의 합이 구해졌다.
            diff_sum += now_diff_sum

    print(f'#{tc} {diff_sum}') 
````

````python
# 풍선팡2

T = int(input())
for tc in range(1, T+1):
    N, M = map(int,input().split)    
    matrix = [list(map(int, input().split())) for _ in range(N)] # 풍선정보

    # 상하좌우로 움직일 때 좌표의 변화량
    dr = [-1,1,0,0]
    dc = [0,0,-1,1]

    # 최대값
    max_cnt = 0

    for r in range(N):

        for c in range(M): # 현재 위치 (r,c) # 현재 위치에 있는 풍선의 꽃가루 개수 = arr[r][c] # 여기에 상하좌우 풍선도 확인해서 꽃가루 개수 # 다 더해준 다음 최대값 구하기
            cnt = 0
            
            for d in range(4): # d는 방향
                nr = r + dr[d] # 중심점 다음 행,열 좌표
                nc = c + dc[d]
                if 0 <= nr < N and 0 <= nc < N: # 배열 내에 있을 때 !

                # 여기까지는 비슷(기본값)
            ###########################################################################################
                # 하고싶은 계산 ㄱㄱ    

                    cnt += matrix[nr][nc]


            max_cnt = cnt if cnt > max_cnt else max_cnt
            max_cnt =max(max_cnt, cnt)

    print(f'#{tc} {max_cnt}')
````
