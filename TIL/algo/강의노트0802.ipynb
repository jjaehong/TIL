#### 2차원 배열 입력받기

arr2 = [list(map(int, input())).split for _ in range(N)] 

    - list() => 한 줄 입력받기 
    - [list() ,for _ in range(N)] => 한 줄을 N번 입력 받아서 2차원 배열로 만든다

#### 2차원 배열의 접근

배열 순회 : n*m 배열의 모든 원소를 빠짐없이 조사하는 방법

1. 행 우선순회
- r행의 좌표, 열의 좌표

```python
for r in range(n):
    for c in range(m):
        f(Array[r][c]) # 필요한 연산 수행
```


2. 열 우선순회
- r행의 좌표, c열의 좌표

```python
for c in range(m):
    # max_v = 0
    # s = 0 : 초기값
    for r in range(n):
        f(Array[r][c]) 

        # if max_v < s 등등

```


3. 지그재그 순회
- r행의 좌표, c열의 좌표

```python
for r in range(n):
    for c in range(m):
        f(Array[r][c + (m-1-2*c)*(r%2)]) 
        # 증가하는 방향 : A[r][c], 
        # 감소하는 방향 : A[r][m-1-c]
        # r % 2 == 0 : c  //// r는 0부터..
        # r % 2 == 1 : (m-1) - c
        # 필요한 연산 수행
```



````python
# 2차원 배열 응용(행 총합의 최댓값)

N = 2
M = 4 
arr = [[0,1,2,3],
       [4,5,6,7]]

max_v = 0 # 최대값 초기화
for i in range(N):
    row_total = 0 # 행 총합 초기화
    for j in range(M):
        row_total += arr[i][j]
    if max_v < row_total:
        max_v = row_total

print(max_v)

````

#### 전치행렬

![이미지](../이미지/전치행렬.PNG)

````python
import sys
sys.stdin = open("input.txt", "r")


T = int(input())
for tc in range(1,T+1):
    N = int(input())
    matrix = [list(map(int, input().split())) for _ in range(N)] # 2차원 배열 입력 받기 # list() 한줄 입력받기 # [ ~ ,for _ in range(N)] => N번 입력 받아서 2차원 배열로 만든다

    # 상하좌우(델타) = (행의좌표, 열의좌표) -> (i,j) (r,c)

    # idx[]상 하 좌 우
    dr = [-1, 1, 0, 0] # 행 번호의 변화량
    dc = [0, 0, -1, 1] # 열 번호의 변화량
    # 상(-1,0) / 하(1,0) / 좌(0,-1) / 우(0,1) 만큼 변한다.


    # 행 우선
    diff_sum = 0
    for r in range(N): # 가로길이 = 세로길이 / 행 = 열 = N
        for c in range(N):
            #print(f'현재위치 : ({r},{c}), 값 : {matrix[r][c]}') # 현재 위치가 r행 c열 (r,c) # matrix[r][c] => 현재 위치에 저장된 값

            now_diff_sum = 0 # 현재 위치에서 상하좌우 차이 절대값의 합

            # 상하좌우 좌표 탐색 
            for d in range(4): # d는 델타 배열의 인덱스[0,1,2,3], d = 0이면 상...
                # 현재 위치는 (r,c) 상하좌우로 이동 후 위치를 (nr,nc)
                nr = r + dr[d] # 다음 행 좌표 = 현재 행 좌표 + 이동방향이 d일 때 변화량
                nc = c + dc[d] # 다음 열 좌표 = 현재 열 좌표 + 이동방향이 d일 때 변화량
                # print(f' 다음 행렬 좌표는 ({nr}, {nc})')



                # 내가 계산한 위치가 유효한 인덱스인지?(배열의 범위 안에 있는지) 검사
                if 0 <= nr < N and 0 <= nc < N:
                    # 차이 = 현재 위치에 있는 값 - 새로 움직인 위치에 있는 값
                    diff = matrix[r][c] - matrix[nr][nc]

                    # 절대값 abs() 사용하지 않고 구하기
                    diff = -diff if diff < 0 else diff
                    now_diff_sum += diff

                # 상하좌우 탐색이 끝나면 현재 위치 절대값 차이의 합이 구해졌다.
            diff_sum += now_diff_sum

    print(f'#{tc} {diff_sum}') 
````

````python
# sum
# 행 총합, 열 총합, 대각선 총합 중 max 값


import sys
sys.stdin = open("input.txt", "r")
for tc in range(1, 11):
    tc = int(input())
    matrix = [list(map(int, input().split())) for _ in range(100)]

    max_row = 0
    max_col = 0
    dia1_tot = 0
    dia2_tot = 0
        
    for r in range(100):
        row_tot = 0
        col_tot = 0
        dia1_tot += matrix[r][r]
        dia2_tot += matrix[r][99-r]


        for c in range(100):
            row_tot += matrix[r][c]
            if max_row < row_tot:
                max_row = row_tot

            col_tot += matrix[c][r]
            if max_col < col_tot:
                max_col = col_tot

    result = [max_row, max_col, dia1_tot, dia2_tot]
    result.sort()

    print(f'#{tc} {result[3]}')
````
